```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, dblquad, tplquad, odeint, solve_ivp

# ================================================================
# 1. Definite Integral (quad)
# ================================================================
print("\n1. Definite Integral (quad):")

def f(x):
    return x**2

result, error = quad(f, 0, 1)  # Integrate x^2 from 0 to 1

print(f"Integral result: {result}")
print(f"Estimated error: {error}")

print("\nConclusion (quad):")
print("quad computes the definite integral of a function of one variable. It's a general-purpose integration function.")

# Use Case: Calculating areas under curves, evaluating probabilities.


# ================================================================
# 2. Double Integral (dblquad)
# ================================================================
print("\n2. Double Integral (dblquad):")

def f(y, x):
    return x*y

result, error = dblquad(f, 0, 1, 0, 2)  # Integrate x*y over a rectangular region

print(f"Double integral result: {result}")
print(f"Estimated error: {error}")

print("\nConclusion (dblquad):")
print("dblquad computes the double integral of a function of two variables. It's used for integrating over 2D regions.")

# Use Case: Calculating volumes, integrating over surfaces.


# ================================================================
# 3. Triple Integral (tplquad)
# ================================================================
print("\n3. Triple Integral (tplquad):")

def f(z, y, x):
    return x*y*z

result, error = tplquad(f, 0, 1, 0, 2, 0, 3) # Integrate x*y*z over a 3D rectangular region

print(f"Triple integral result: {result}")
print(f"Estimated error: {error}")

print("\nConclusion (tplquad):")
print("tplquad computes the triple integral of a function of three variables. It's used for integrating over 3D volumes.")

# Use Case: Calculating mass, center of mass, integrating over volumes.


# ================================================================
# 4. Ordinary Differential Equations (odeint)
# ================================================================
print("\n4. Ordinary Differential Equations (odeint):")

def dydt(y, t):
    return -y  # Example: dy/dt = -y

y0 = 1  # Initial condition
t = np.linspace(0, 5, 100)  # Time points

y = odeint(dydt, y0, t)

plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('ODE Solution (odeint)')
plt.show()

print("\nConclusion (odeint):")
print("odeint solves a system of first-order ordinary differential equations (ODEs). It's used for modeling dynamic systems.")

# Use Case: Physics simulations, population dynamics, control systems.


# ================================================================
# 5. Initial Value Problem Solver (solve_ivp)
# ================================================================
print("\n5. Initial Value Problem Solver (solve_ivp):")

def dydt(t, y):
    return -y  # Example: dy/dt = -y

y0 = [1]  # Initial condition (must be a list or array)
t_span = (0, 5)  # Time span
t_eval = np.linspace(0, 5, 100) # Time points to evaluate the solution

sol = solve_ivp(dydt, t_span, y0, t_eval=t_eval)

plt.plot(sol.t, sol.y[0]) # sol.y is a 2D array
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('ODE Solution (solve_ivp)')
plt.show()

print("\nConclusion (solve_ivp):")
print("solve_ivp solves initial value problems for ODEs. It offers more options and control than odeint, including handling stiff systems and events.")

# Use Case: Physics simulations, control theory, more complex dynamic systems.
```
