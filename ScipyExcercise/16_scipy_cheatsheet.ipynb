# **âš”ï¸ NumPy vs SciPy â€“ Quick Comparison**

| Feature / Use Case               | âœ… NumPy                             | ğŸ”¬ SciPy                                 |
|-------------------------------|-------------------------------------|-----------------------------------------|
| ğŸ“¦ **Purpose**                 | Core array & math operations        | Built on top of NumPy â€“ advanced tools  |
| ğŸ§® **Array Support**           | âœ… `ndarray` is core                 | ğŸš« Uses NumPy arrays internally          |
| â• **Basic Math & Algebra**     | Yes (e.g., `np.sum`, `np.dot`)      | Yes (but via `scipy.linalg`)            |
| ğŸ“Š **Statistics**             | Basic (mean, std, var)              | Advanced (`scipy.stats`)                |
| ğŸ“ **Linear Algebra**          | Basic (`np.linalg`)                 | Advanced (`scipy.linalg`)               |
| ğŸ“ **Optimization**            | âŒ Not available                     | âœ… `scipy.optimize`                      |
| ğŸ“š **Distributions**           | âŒ Not available                     | âœ… `scipy.stats` (PDF, CDF, fitting)     |
| ğŸ” **Interpolation**           | Basic via `np.interp`               | Advanced (`scipy.interpolate`)          |
| ğŸ§ª **Numerical Integration**   | âŒ Not available                     | âœ… `scipy.integrate`                     |
| ğŸ§Š **Image Processing**        | âŒ Not available                     | âœ… `scipy.ndimage`                       |
| ğŸ§  **Clustering & Distance**   | âŒ Not available                     | âœ… `scipy.cluster`, `scipy.spatial`     |
| ğŸ§¾ **Signal Processing**       | Very limited                        | âœ… `scipy.signal`                        |
| ğŸ“¥ **Dependencies**            | Lightweight                         | Heavier (depends on NumPy)              |

---

### âœ… Summary:
- **NumPy** = ğŸ”¢ Fast array ops + math foundations  
- **SciPy** = ğŸ§  Advanced scientific tools **built on NumPy**  
- **Use together** in almost every ML/DS pipeline!

### 4ï¸âƒ£ Optimize Model Hyperparameters Manually  
ğŸ“Œ Use `scipy.optimize.minimize` for custom tuning  
```python
from scipy.optimize import minimize
from sklearn.model_selection import cross_val_score

def loss_fn(params):
    model = KNeighborsClassifier(n_neighbors=int(params[0]))
    return -cross_val_score(model, X, y, cv=3).mean()

result = minimize(loss_fn, [3], bounds=[(1, 10)])
best_k = int(result.x[0])
```
