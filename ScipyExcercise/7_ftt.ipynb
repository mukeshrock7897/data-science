```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq, rfft, irfft, fftshift

# ================================================================
# 1. Fast Fourier Transform (fft)
# ================================================================
print("\n1. Fast Fourier Transform (fft):")

fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, False)  # 1 second
x = np.sin(2*np.pi*100*t) + 0.5*np.sin(2*np.pi*200*t)  # Two sine waves

xf = fft(x)
xf_abs = np.abs(xf) #Magnitude of the FFT
freqs = fftfreq(x.size, d=1/fs) #Frequencies corresponding to the FFT output

plt.plot(freqs[:fs//2], xf_abs[:fs//2])  # Plot positive frequencies only
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT of Two Sine Waves")
plt.show()

print("\nConclusion (fft):")
print("fft computes the Discrete Fourier Transform (DFT) of a signal using the Fast Fourier Transform algorithm. It decomposes a signal into its constituent frequencies.")

# Use Case: Spectral analysis, frequency domain filtering, audio processing.

# ================================================================
# 2. Inverse Fast Fourier Transform (ifft)
# ================================================================
print("\n2. Inverse Fast Fourier Transform (ifft):")

xf = fft(x)  # From the previous example
x_reconstructed = ifft(xf)

print(f"Original signal (first 5 values): {x[:5]}")
print(f"Reconstructed signal (first 5 values): {x_reconstructed[:5]}")

print("\nConclusion (ifft):")
print("ifft computes the inverse DFT, reconstructing the original time-domain signal from its frequency components.")

# Use Case: Signal reconstruction, synthesis.

# ================================================================
# 3. FFT Frequencies (fftfreq)
# ================================================================
print("\n3. FFT Frequencies (fftfreq):")

freqs = fftfreq(x.size, d=1/fs)

print(f"Frequencies (first 10 values): {freqs[:10]}")

print("\nConclusion (fftfreq):")
print("fftfreq generates the frequencies corresponding to the output of fft. It's essential for correctly interpreting the frequency spectrum.")

# Use Case: Labeling frequency axes, spectral analysis.

# ================================================================
# 4. Real FFT (rfft)
# ================================================================
print("\n4. Real FFT (rfft):")

xf_real = rfft(x)
freqs_real = np.fft.rfftfreq(x.size, d=1/fs) #Corresponding frequencies

plt.plot(freqs_real, np.abs(xf_real))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("RFFT of Two Sine Waves")
plt.show()


print("\nConclusion (rfft):")
print("rfft computes the DFT of a real-valued signal. It's more efficient than fft for real signals because it only computes the positive frequencies (the negative frequencies are redundant for real signals).")

# Use Case: Spectral analysis of real-valued signals, audio processing.

# ================================================================
# 5. Inverse Real FFT (irfft)
# ================================================================
print("\n5. Inverse Real FFT (irfft):")

x_real_reconstructed = irfft(xf_real)

print(f"Original signal (first 5 values): {x[:5]}")
print(f"Reconstructed real signal (first 5 values): {x_real_reconstructed[:5]}")

print("\nConclusion (irfft):")
print("irfft computes the inverse DFT of a real-valued signal's frequency spectrum, reconstructing the original time-domain signal.")

# Use Case: Signal reconstruction from real FFT coefficients.

# ================================================================
# 6. FFT Shift (fftshift)
# ================================================================
print("\n6. FFT Shift (fftshift):")

xf_shifted = fftshift(xf)
freqs_shifted = fftshift(freqs)

plt.plot(freqs_shifted, np.abs(xf_shifted))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT Shifted")
plt.show()

print("\nConclusion (fftshift):")
print("fftshift rearranges the FFT output so that the zero-frequency component is in the center of the spectrum. This makes it easier to visualize the frequency content of a signal, especially when dealing with both positive and negative frequencies.")

# Use Case: Visualizing frequency spectra, spectral analysis.
```
