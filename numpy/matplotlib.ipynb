```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("/Users/apple/Documents/GitHub/ML/numpy/Automobile_data.csv")
dataset.dropna(inplace=True)

#converting float64 into int64 these below columns
dataset['length']=dataset['length'].astype('int64')
dataset['wheel-base']=dataset['wheel-base'].astype('int64')
dataset['price']=dataset['price'].astype('int64')

# Select numerical columns for the heatmap
numerical_columns = ['wheel-base', 'length' ,'horsepower', 'average-mileage', 'price']


selected_data = dataset[numerical_columns]

# Create the heatmap using correlation matrix
correlation = selected_data.corr()

# Create a new figure for plotting
plt.figure(figsize=(10, 6))

# Generate the heatmap with colormap and annotation
plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')

# Add labels for each cell (optional)
tick_labels = correlation.columns
plt.xticks(range(len(tick_labels)), tick_labels, rotation=45, ha='right')
plt.yticks(range(len(tick_labels)), tick_labels)

# # Annotate heatmap with correlation values
for i in range(len(correlation.columns)):
  for j in range(len(correlation.columns)):
    plt.text(i, j, "{:.2f}".format(correlation.iloc[i, j]), ha='center', va='center')

plt.title('Correlation Heatmap of Selected Numerical Features')
plt.tight_layout()
plt.show()
```

```python
# creating voilin plots 

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("/Users/apple/Documents/GitHub/ML/numpy/Automobile_data.csv")
dataset.dropna(inplace=True)

#converting float64 into int64 these below columns
dataset['length']=dataset['length'].astype('int64')
dataset['wheel-base']=dataset['wheel-base'].astype('int64')
dataset['price']=dataset['price'].astype('int64')


# Select numerical columns for the violin plot
numerical_columns = ['wheel-base', 'length', 'horsepower', 'average-mileage', 'price']

# Create a violin plot for each numerical column
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns):
    plt.subplot(2, 3, i + 1)
    plt.violinplot(dataset[column])
    plt.title(f"{column}")

# Adjust spacing and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("/Users/apple/Documents/GitHub/ML/numpy/Automobile_data.csv")
dataset.dropna(inplace=True)

#converting float64 into int64 these below columns
dataset['length']=dataset['length'].astype('int64')
dataset['wheel-base']=dataset['wheel-base'].astype('int64')
dataset['price']=dataset['price'].astype('int64')

# Select numerical columns for the hexbin plot
numerical_columns = ['wheel-base', 'length', 'horsepower', 'average-mileage', 'price']

# Create a hexbin plot for each pair of numerical columns
fig, axes = plt.subplots(nrows=len(numerical_columns), ncols=len(numerical_columns), figsize=(20, 20))

for i, column1 in enumerate(numerical_columns):
    for j, column2 in enumerate(numerical_columns):
        if i != j:  # Avoid plotting against itself
            axes[i, j].hexbin(dataset[column1], dataset[column2], gridsize=40, cmap='viridis')
            axes[i, j].set_xlabel(column1)
            axes[i, j].set_ylabel(column2)

# Adjust spacing and show the plot
plt.tight_layout()
plt.show()
```
