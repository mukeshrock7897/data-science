```python
#lineplot

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [1000, 1200, 1500, 1800, 2000, 2200]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a line plot
sns.lineplot(x='Month', y='Sales', data=df)

# Add title and labels
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

```python
#voilinplot 

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [1000, 1200, 1500, 1800, 2000, 2200]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert 'Month' to categorical data
df['Month'] = pd.Categorical(df['Month'])

# Create a violin plot
sns.violinplot(x='Month', y='Sales', data=df)

# Add title and labels (optional)
plt.title('Monthly Sales Violin Plot')
plt.xlabel('Month')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

### Distribution plots

```python
#regplot

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {
    'Marketing_Spend': [10000, 12000, 15000, 18000, 20000],
    'Sales': [20000, 22000, 25000, 28000, 30000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a regression plot
sns.regplot(x='Marketing_Spend', y='Sales', data=df)

# Add title and labels
plt.title('Sales vs. Marketing Spend')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')

# Show the plot
plt.show()
```
