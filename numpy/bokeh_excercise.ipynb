```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook  # For Jupyter Notebook output
from bokeh.models import ColumnDataSource, Circle, Line, Title, Label  # Import glyph classes and layout elements

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a ColumnDataSource to manage data
source = ColumnDataSource(data=dict(x=x, y=y))

# Create a figure with desired title, width, and height
p = figure(title="Enhanced Line Plot with Glyphs and Layout", width=600, height=400)

# Add a line glyph with data source and styling
p.line('x', 'y', source=source, line_width=2, color="red")

# Add a circle glyph at specific points (optional)
circle_glyph = Circle(x='x', y='y', radius=15, fill_color="yellow", line_color="red")
p.add_glyph(source, circle_glyph)  # Add the circle glyph to the plot

# Add interactive tools for panning and zooming (optional)
p.add_tools(
    "pan",  # Panning tool
    "zoom_in", "zoom_out", "reset",  # Zooming and reset tools
)

# Add layout elements (optional)
title = Title(text="Custom Plot Title", align="center")
p.add_layout(title, "above")  # Add title above the plot

label = Label(x=4, y=6, text="Data Point Label", text_align="center", text_font="bold")
p.add_layout(label)  # Add a label to a specific point

# Optionally add layout elements like axes or grids (uncomment if desired)
# p.xaxis.axis_label = "X-Axis"  # Set axis labels
# p.yaxis.axis_label = "Y-Axis"
# p.grid.grid_line_color = None  # Remove grid lines for cleaner visuals (optional)

# Output the plot to a Jupyter Notebook (if applicable)
output_notebook()

# Display the plot
show(p)
```

```python
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

data = {'fruits': ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries'],
        'counts': [5, 3, 4, 2, 4, 6]}
source = ColumnDataSource(data=data)

# changed plot_height to height
p = figure(x_range=data['fruits'], height=750, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.5, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=data['fruits']))

p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```

```python

from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
data = np.random.rand(10, 2)

# Perform hierarchical clustering
Z = linkage(data, method='ward')

# Plot dendrogram
plt.figure()
dendrogram(Z)
plt.show()
```
