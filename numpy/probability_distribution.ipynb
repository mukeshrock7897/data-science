### **Probability Distribution**
* A probability distribution is a mathematical function that describes the probability of different possible values of a variable.
* A probability distribution describes how the values of a random variable are distributed.
* It provides a mathematical function that gives the probabilities of occurrence of different possible outcomes for an experiment.


### **Key Concepts**

### **1. Random Variable**

* A random variable is a variable whose value depends on the outcome of a probabilistic experiment
* It can be either discrete or continuous.
* in discrete where we get countable numbers like rolling the dice
* in continuous where we get the continuous nature values (any value within a continuous range) like time

### **2.Probabilty Functions**
* Probability functions are mathematical representations that describe the likelihood of different possible outcomes of a random variable

### 1. **Definition of Probability Mass Function (PMF)**:
The **Probability Mass Function (PMF)** is a function that gives the probability that a discrete random variable is exactly equal to some value. It defines the probability distribution of a discrete random variable and assigns probabilities to individual outcomes.

### 3. **Where We Use PMF**:
PMF is used in cases where we are dealing with **discrete random variables**, meaning the variable can only take distinct, countable values. Common use cases include:
- Probability of rolling a specific number on a die.
- The number of heads in multiple flips of a coin.
- Counting the number of successes in a fixed number of independent trials (binomial experiments).

### 4. **When to Use PMF**:
Use the PMF when:
- The variable is **discrete** (i.e., it has countable outcomes like integers or categories).
- You need to compute the exact probability of specific outcomes.
- You are dealing with probability models for events like coin flips, die rolls, or the number of occurrences of an event (e.g., Poisson, binomial distributions).

### 5. **Real-Time Industry Examples**:
- **Manufacturing**: Estimating the probability of producing a defective product in a batch of 100 units (e.g., binomial distribution).
- **Marketing**: Finding the probability that a customer will make exactly 2 purchases in a week based on historical data (e.g., Poisson distribution).
- **Finance**: Modeling the probability that an insurance claim will occur for a discrete number of policyholders within a specific time period.
- **Gaming**: Calculating the probability of a specific outcome (like rolling a 6) in board games involving dice.
