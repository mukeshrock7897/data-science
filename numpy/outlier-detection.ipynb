```python
import numpy as np
import matplotlib.pyplot as plt

# Corrected dataset
x = [2, 4, 6, 8, 10, 12, 14]
y = [4, 6, 36, 100, 144, 400, 4000]

# Creating scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter plot of X vs Y')
plt.grid(True)

# Detecting outliers using IQR for Y values
q1, q3 = np.percentile(y, [25, 75])
iqr = q3 - q1

lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

# Outliers are values beyond the lower or upper fences
outliers = [val for val in y if val < lower_fence or val > upper_fence]

# Highlighting outliers in the plot
for i, val in enumerate(y):
    if val in outliers:
        plt.scatter(x[i], y[i], color='red', label='Outlier' if outliers.index(val) == 0 else "")

# Displaying the scatter plot with outlier legend
plt.legend()
plt.show()

# Printing the outliers
print(f"Outliers: {outliers}")
```

```python
# Z-Score
outliers=[]

def DetectOutliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        zscore=(i-mean)/std
        if np.abs(zscore)>threshold:
            outliers.append(i)

    return outliers

 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example Dataset with an outlier
dataset = [5, 7, 8, 12, 13, 14, 16, 18, 21, 22, 23, 25, 27, 29, 30, 100]  # Added 100 as an outlier

# Calculate five-number summary
minimum = np.min(dataset)
q1 = np.percentile(dataset, 25)  # 1st quartile
median = np.median(dataset)        # Median (2nd quartile)
q3 = np.percentile(dataset, 75)   # 3rd quartile
maximum = np.max(dataset)
five_number_summary = [minimum, q1, median, q3, maximum]

# Calculate IQR and identify outliers
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
outliers = [x for x in dataset if x < lower_fence or x > upper_fence]

# Create the boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(dataset, vert=False)
plt.xlabel('Values')

# Annotate the boxplot with component values
plt.text(minimum - 5, 1, f'Min: {minimum}', horizontalalignment='right', fontsize=10)
plt.text(q1 - 5, 1, f'Q1: {q1}', horizontalalignment='right', fontsize=10)
plt.text(median - 5, 1, f'Median: {median}', horizontalalignment='right', fontsize=10)
plt.text(q3 + 1, 1, f'Q3: {q3}', horizontalalignment='left', fontsize=10)
plt.text(maximum + 1, 1, f'Max: {maximum}', horizontalalignment='left', fontsize=10)

# Highlight the outliers in the boxplot
if outliers:
    for outlier in outliers:
        plt.scatter(outlier, 1, color='red', label='Outlier')

# Add title and display the boxplot
plt.title('Boxplot with Five Number Summary Annotations')
plt.grid(axis='x')
plt.show()
```
