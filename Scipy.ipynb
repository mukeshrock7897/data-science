# ðŸ“¦ `scipy.stats` â€“ Statistics & Distributions

### âœ… What it is:
`scipy.stats` provides tools for **descriptive statistics**, **statistical tests**, **probability distributions**, and **correlations** â€” used widely in **data science, machine learning, and research**.

---

## 1ï¸âƒ£ Descriptive Statistics

### ðŸ‘‰ Purpose:
Quick summary of your data â€” central tendency, spread, and shape.

| Function       | Description                            |
|----------------|----------------------------------------|
| `mean()`       | Average of values                      |
| `median()`     | Middle value                           |
| `std()`        | Standard deviation                     |
| `var()`        | Variance                               |
| `mode()`       | Most frequent value                    |
| `skew()`       | Asymmetry of the distribution          |
| `kurtosis()`   | Peakedness or flatness of distribution |

### ðŸ§  When to Use:
- During **Exploratory Data Analysis (EDA)** to understand your dataset.

### ðŸ“› When NOT to Use:
- If your data has **missing values**, clean it first.

### âœ… Code Example:
```python
from scipy import stats
import numpy as np

data = [1, 2, 2, 3, 4, 4, 4, 5, 6]

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data, keepdims=True).mode[0])
print("Standard Deviation:", np.std(data))
print("Skewness:", stats.skew(data))
print("Kurtosis:", stats.kurtosis(data))
```

---

## 2ï¸âƒ£ Statistical Tests

### ðŸ‘‰ Purpose:
Used to check if assumptions/statements about data are true (hypothesis testing).

| Function         | Description                                        | Use Case                                 |
|------------------|----------------------------------------------------|-------------------------------------------|
| `ttest_ind()`     | Compare two independent groups                    | A/B testing, gender-based comparison      |
| `ttest_rel()`     | Compare two related samples                       | Before/after experiment                   |
| `chisquare()`     | Categorical data goodness-of-fit                  | Are observed frequencies expected?        |
| `ks_2samp()`      | Compare two distributions                         | Are two datasets similarly distributed?   |
| `shapiro()`       | Test for normal distribution                      | Normality check before t-test             |
| `mannwhitneyu()`  | Non-parametric test for independent groups        | Skewed data comparison                    |

### âœ… Code Example:
```python
# Independent t-test
a = [5, 6, 7, 8]
b = [6, 7, 8, 9]
t_stat, p_val = stats.ttest_ind(a, b)
print("T-Statistic:", t_stat, "| P-Value:", p_val)
```

### ðŸ§  When to Use:
- To **prove** or **disprove** a hypothesis using data.

### ðŸ“› When NOT to Use:
- Donâ€™t apply parametric tests (like t-tests) if data is **not normally distributed**.

---

## 3ï¸âƒ£ Correlation

### ðŸ‘‰ Purpose:
Measure the **strength and direction** of relationship between two variables.

| Function       | Measures                 | Type of Data          |
|----------------|--------------------------|------------------------|
| `pearsonr()`   | Linear correlation       | Continuous, linear     |
| `spearmanr()`  | Rank-based correlation   | Non-linear or ordinal  |
| `kendalltau()` | Rank-based, robust       | Small datasets         |

### âœ… Code Example:
```python
x = [10, 20, 30, 40, 50]
y = [12, 24, 33, 47, 55]

print("Pearson:", stats.pearsonr(x, y))
print("Spearman:", stats.spearmanr(x, y))
```

### ðŸ§  When to Use:
- To test if one feature is related to another.
- Feature selection in ML.

---

## 4ï¸âƒ£ Distributions

### ðŸ‘‰ Purpose:
Used to model data probabilistically â€” sample data, fit data, get probabilities.

### ðŸ“Œ 4.1 Continuous Distributions

| Function     | Description                       |
|--------------|-----------------------------------|
| `norm`       | Normal (Gaussian) distribution    |
| `expon`      | Exponential (decay events)        |
| `gamma`, `beta` | Shape-based continuous models |

#### âœ… Code Example:
```python
from scipy.stats import norm

# Probability density function (PDF)
print("PDF at x=0:", norm.pdf(0, loc=0, scale=1))

# Cumulative distribution function (CDF)
print("CDF at x=1:", norm.cdf(1))

# Random samples
samples = norm.rvs(loc=0, scale=1, size=5)
print("Samples:", samples)
```

---

### ðŸ“Œ 4.2 Discrete Distributions

| Function     | Description                 |
|--------------|-----------------------------|
| `binom`      | Binary outcomes (success/fail) |
| `poisson`    | Count of events in interval    |

#### âœ… Code Example:
```python
from scipy.stats import binom

# Probability of exactly 2 successes in 5 trials
print("P(X=2):", binom.pmf(2, n=5, p=0.5))
```

---

### ðŸ“Œ 4.3 Multivariate Distributions

| Function              | Description                      |
|-----------------------|----------------------------------|
| `multivariate_normal` | Multi-feature Gaussian model     |
| `dirichlet`           | Probabilities that sum to 1      |

#### âœ… Code Example:
```python
from scipy.stats import multivariate_normal
mean = [0, 0]
cov = [[1, 0], [0, 1]]

rv = multivariate_normal(mean, cov)
print("PDF:", rv.pdf([0, 0]))
```

---

## 5ï¸âƒ£ KDE (Kernel Density Estimation)

### ðŸ‘‰ Purpose:
Estimate smooth distribution from data (non-parametric). Useful for visualization.

| Function         | Description                        |
|------------------|------------------------------------|
| `gaussian_kde()` | Estimate PDF from sample data      |

#### âœ… Code Example:
```python
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

data = [1, 2, 2, 3, 4, 5, 6]
kde = gaussian_kde(data)
x = np.linspace(0, 7, 100)
plt.plot(x, kde(x))
plt.title("KDE Plot")
plt.show()
```

---

## ðŸ“Š When to Use vs. Not Use

| When to Use                                          | When NOT to Use                                  |
|------------------------------------------------------|--------------------------------------------------|
| For EDA, checking normality, comparing groups        | If data is biased, missing, or needs transformation |
| Modeling or fitting data to distributions            | Donâ€™t use t-tests blindly on skewed data         |
| Feature correlation, hypothesis testing              | Donâ€™t use KDE if data is too small               |

# ðŸ”§ `scipy.optimize` â€“ Optimization & Curve Fitting

### âœ… What it is:
This module helps you **optimize** (minimize or maximize) a function, **fit curves**, **solve equations**, or **optimize parameters** under constraints â€” a must for **ML training**, **hyperparameter tuning**, and **engineering problems**.

---

## 1ï¸âƒ£ Minimization

### ðŸ‘‰ Purpose:
Find the **minimum value** of a function (cost, error, loss).

### ðŸ“Œ Common Functions:
| Function                | Description                              |
|-------------------------|------------------------------------------|
| `minimize()`            | General-purpose minimizer for multivariable functions |
| `minimize_scalar()`     | Minimizer for single-variable functions  |
| `basinhopping()`        | Global optimization (avoids local min)   |
| `differential_evolution()` | Evolutionary algorithm, global search |

### âœ… Code Example:
```python
from scipy.optimize import minimize

def cost(x):
    return x[0]**2 + x[1]**2  # simple bowl-shaped function

result = minimize(cost, x0=[1, 2])
print("Minimum at:", result.x)
```

### ðŸ“˜ Parameters:
- `fun`: the function to minimize
- `x0`: starting point (guess)
- `method`: optimization method (`'BFGS'`, `'Nelder-Mead'`, etc.)
- `bounds`, `constraints`: for restrictions

### ðŸ§  When to Use:
- Cost/loss function minimization
- Model training with custom error functions

### ðŸ“› When NOT to Use:
- For non-smooth functions without global search (`basinhopping` or `differential_evolution` is better)

---

## 2ï¸âƒ£ Root Finding

### ðŸ‘‰ Purpose:
Find **where a function equals zero** (f(x) = 0)

### ðŸ“Œ Common Functions:
| Function        | Use When                |
|-----------------|-------------------------|
| `root()`        | Multivariate equations  |
| `fsolve()`      | Similar to `root`, but older |
| `root_scalar()` | One variable equation   |

### âœ… Code Example:
```python
from scipy.optimize import root

def eq(x):
    return x**2 - 4

result = root(eq, x0=1)
print("Root:", result.x[0])  # Should be close to 2
```

### ðŸ§  When to Use:
- Equation solving: `f(x) = 0`
- In scientific computing models (e.g. solving balance equations)

### ðŸ“› When NOT to Use:
- If function has multiple roots and you want them all â€” this gives **one root only**

---

## 3ï¸âƒ£ Curve Fitting

### ðŸ‘‰ Purpose:
Fit a curve to your data by adjusting parameters.

### ðŸ“Œ Common Functions:
| Function         | Use Case                          |
|------------------|------------------------------------|
| `curve_fit()`    | Non-linear curve fitting (easy!)   |
| `least_squares()`| Fit function with more control     |

### âœ… Code Example:
```python
import numpy as np
from scipy.optimize import curve_fit

def model(x, a, b):
    return a * x + b

x_data = np.array([1, 2, 3, 4])
y_data = np.array([2.1, 4.1, 6.2, 8.1])

params, _ = curve_fit(model, x_data, y_data)
print("a, b:", params)  # approx: a â‰ˆ 2, b â‰ˆ 0.1
```

### ðŸ§  When to Use:
- To fit data (regression, trend analysis)
- Calibration problems

### ðŸ“› When NOT to Use:
- If your model is wrong â€” fitting will be meaningless

---

## 4ï¸âƒ£ Linear Programming

### ðŸ‘‰ Purpose:
Solve optimization problems with **linear** objective functions and constraints.

### ðŸ“Œ Function:
- `linprog()`: minimizes a linear function `cáµ€x` subject to bounds and constraints.

### âœ… Code Example:
```python
from scipy.optimize import linprog

# Minimize: f = x + y
# Subject to: x + 2y â‰¥ 4, 3x + y â‰¥ 3
c = [1, 1]  # Coefficients of x and y in the objective
A = [[-1, -2], [-3, -1]]
b = [-4, -3]

result = linprog(c, A_ub=A, b_ub=b)
print("x, y:", result.x)
```

### ðŸ§  When to Use:
- Resource allocation
- Supply chain, production planning

### ðŸ“› When NOT to Use:
- Nonlinear constraints or objectives â†’ use `minimize` instead

---

## 5ï¸âƒ£ Constrained Optimization

### ðŸ‘‰ Purpose:
Optimize complex functions with **bounds or constraints** (nonlinear or linear).

### ðŸ“Œ Functions:
| Function           | Use Case                                 |
|--------------------|-------------------------------------------|
| `shgo()`           | Global optimization with constraints      |
| `dual_annealing()` | Global optimization using annealing       |

### âœ… Code Example:
```python
from scipy.optimize import dual_annealing

def func(x):
    return x[0]**2 + 5 * np.sin(x[0])

bounds = [(-10, 10)]
result = dual_annealing(func, bounds)
print("Best x:", result.x)
```

---

## âš™ï¸ When to Use vs Not Use

| âœ… Use When...                                 | âŒ Don't Use When...                          |
|------------------------------------------------|-----------------------------------------------|
| You need to minimize loss, error, cost         | Function is not defined (discontinuities)     |
| Solve equations numerically                    | You need symbolic solution (use SymPy)        |
| Fit curve or estimate parameters from data     | The model is wrong or data is noisy           |
| Solve constrained optimization (e.g., LP)      | Problem is non-convex with many discontinuities|

---

# ðŸ§® `scipy.linalg` â€“ Linear Algebra

### âœ… What it is:

---

## ðŸ”¹ 1. Matrix Solvers

### âœ… Purpose:
Solve equations like **Ax = b** or estimate solutions when exact ones donâ€™t exist.

### ðŸ”§ Functions:
| Function | Description |
|----------|-------------|
| `solve(A, b)` | Solves Ax = b |
| `lstsq(A, b)` | Least squares solution (if A is not square or exact solution doesnâ€™t exist) |

### âœ… Code:
```python
from scipy.linalg import solve, lstsq
import numpy as np

A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])

x = solve(A, b)
print("Solution:", x)  # Exact

x_ls, *_ = lstsq(A, b)
print("Least squares:", x_ls)  # Approx
```

### ðŸ§  Use when:
- Solving linear equations
- Fitting linear models

### ðŸš« Donâ€™t use when:
- A is singular (use `lstsq` instead of `solve`)

## ðŸ”¹ 2. Matrix Operations

### âœ… Purpose:
Basic matrix properties & operations.

### ðŸ”§ Functions:
| Function | Meaning |
|----------|---------|
| `inv(A)` | Inverse of A |
| `det(A)` | Determinant |
| `norm(A)` | Norm (size/length of matrix) |
| `trace(A)` | Sum of diagonal elements |

### âœ… Code:
```python
from scipy.linalg import inv, det, norm, trace

A = np.array([[4, 2], [3, 1]])

print("Inverse:\n", inv(A))
print("Determinant:", det(A))
print("Norm:", norm(A))
print("Trace:", trace(A))
```

### ðŸ§  Use when:
- Needed in ML/DS math (e.g. normalization, variance, optimization)

### ðŸš« Donâ€™t use:
- `inv()` directly in solving equations (prefer `solve()` â€“ it's faster and safer)

## ðŸ”¹ 3. Matrix Decompositions

### âœ… Purpose:
Break down complex matrices into simpler parts to understand them or speed up computation.

### ðŸ”§ Functions & Use:

| Function     | Use For                        |
|--------------|--------------------------------|
| `svd(A)`     | Singular Value Decomposition (PCA, dimensionality reduction) |
| `eig(A)`     | Eigenvalues/eigenvectors (PCA, system dynamics) |
| `eigh(A)`    | Same as `eig()` but for symmetric matrices |
| `qr(A)`      | QR decomposition (used in solving) |
| `lu(A)`      | LU decomposition (solve Ax = b) |
| `cholesky(A)`| Fast solve for symmetric positive definite matrices |

### âœ… Code:
```python
from scipy.linalg import svd, eig, qr

A = np.array([[1, 2], [3, 4]])

U, s, Vh = svd(A)
print("SVD â€“ U:", U, "s:", s, "Vh:", Vh)

vals, vecs = eig(A)
print("Eigenvalues:", vals)
print("Eigenvectors:\n", vecs)

Q, R = qr(A)
print("QR â€“ Q:\n", Q, "\nR:\n", R)
```

### ðŸ§  Use when:
- Doing PCA, data compression, understanding model internals
- Speeding up large matrix calculations

### ðŸš« Donâ€™t use:
- On extremely large data without batching or GPU (performance issues)

# ðŸŒ± `scipy.sparse` â€“ Sparse Matrices

### âœ… What it is:
`sparse` helps **store and process big matrices with lots of zeros efficiently**, using less memory and time.

---

## ðŸ§  Why Sparse?
- Many real-world datasets (like **TF-IDF**, **graphs**, **recommendations**) have huge matrices mostly filled with **0s**.
- Dense matrices waste memory for these.
- Sparse matrices only store **non-zero values**.

---

## ðŸ”¹ 1. Sparse Matrix Formats

### âœ… Purpose:
Choose the best structure to store and operate on sparse data.

### ðŸ”§ Common Formats:

| Format         | Best For                          | Usage                             |
|----------------|------------------------------------|-----------------------------------|
| `csr_matrix`   | Row operations (fast `.dot()`)     | Compressed Sparse Row             |
| `csc_matrix`   | Column operations                  | Compressed Sparse Column          |
| `coo_matrix`   | Easy construction (row, col, data) | Coordinate format                 |
| `dok_matrix`   | Easy incremental building          | Dict of Keys format               |
| `bsr_matrix`   | Block matrices (used in systems)   | Block Sparse Row                  |

### âœ… Code Example:
```python
from scipy.sparse import csr_matrix, coo_matrix
import numpy as np

dense = np.array([[0, 0, 3], [4, 0, 0], [0, 5, 0]])

csr = csr_matrix(dense)
print(csr)
```

---

## ðŸ”¹ 2. Matrix Creation

### ðŸ”§ Functions:

| Function     | Description                   |
|--------------|-------------------------------|
| `eye(n)`     | Identity matrix (sparse)       |
| `identity(n)`| Identity (like `np.identity`)  |
| `spdiags()`  | Create matrix from diagonals   |

### âœ… Code:
```python
from scipy.sparse import eye, identity, spdiags

I = identity(3)
print("Identity matrix:\n", I.toarray())

data = np.array([[1, 2, 3]])
diagonals = spdiags(data, [0], 3, 3)
print("Diagonal matrix:\n", diagonals.toarray())
```

---

## ðŸ”¹ 3. Sparse Matrix Operations

### ðŸ”§ Common Ops:

| Operation     | Use                          |
|---------------|-------------------------------|

## **scipy.cluster â€“ Clustering in SciPy**

# ðŸ¤ `scipy.cluster` â€“ Clustering in SciPy

### ðŸ“Œ *Used For: Grouping similar data points (unsupervised learning)*

---

## ðŸ”· 1. Hierarchical Clustering

### âœ… What:
Builds a **tree of clusters** (dendrogram) based on **distance similarity** between points.

---

### ðŸ”§ Common Functions:

| Function       | Description |
|----------------|-------------|
| `linkage()`     | Computes the hierarchy (tree) of clusters |
| `dendrogram()`  | Plots the cluster tree |
| `fcluster()`    | Forms flat clusters from the hierarchy |
| `cophenet()`    | Correlation between original distances and tree |
| `inconsistent()`| Measures inconsistency of each node |

---

### âœ… Example:
```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1, 2], [1, 4], [5, 6], [8, 8]])

# Create linkage matrix (method: 'ward', 'single', 'complete', etc.)
Z = linkage(X, method='ward')

# Plot dendrogram
dendrogram(Z)
plt.title("Dendrogram")
plt.show()

# Form flat clusters (cut at t=7)
clusters = fcluster(Z, t=7, criterion='distance')
print("Cluster labels:", clusters)
```

### âœ… Parameters:

- `metric`: Distance metric (default: `'euclidean'`)
- `t`: Threshold to cut dendrogram in `fcluster`
- `criterion`: How to decide clusters (e.g., `'distance'`, `'maxclust'`)

---

### ðŸ§  When to Use:
- Small to medium-sized datasets  
- You want to visualize clustering structure  
- You donâ€™t need to predefine the number of clusters  

### âŒ When Not to Use:
- Very large datasets (slow and memory-heavy)  
- Need scalable clustering (use KMeans or DBSCAN instead)

## ðŸ”· 2. K-Means & Vector Quantization

### âœ… What:
Divides data into **k groups (clusters)** by minimizing within-cluster distance.

---

### ðŸ”§ Common Functions:

| Function    | Description |
|-------------|-------------|
| `kmeans()`   | Run basic K-means clustering |
| `kmeans2()`  | Improved version with more options |
| `vq()`       | Assigns each point to a cluster (vector quantization) |
| `whiten()`   | Normalize each feature to unit variance |

---

### âœ… Example:
```python
from scipy.cluster.vq import kmeans, vq, whiten
import numpy as np

# Data
X = np.array([[1, 2], [1, 4], [5, 6], [8, 8]])

# Normalize features
X_whitened = whiten(X)

# Run KMeans (k=2)
centroids, distortion = kmeans(X_whitened, 2)

# Assign each point to a cluster
labels, _ = vq(X_whitened, centroids)

print("Cluster Centers:\n", centroids)
print("Cluster Labels:", labels)
```

---

### âœ… Parameters:

- `kmeans(X, k)`:  
  - `X`: Data array  
  - `k`: Number of clusters  
- `vq(X, centroids)`:  
  - Assigns each row in `X` to closest centroid  
- `whiten(X)`:  
  - Scales each feature to have unit variance (improves clustering accuracy)

---

### ðŸ§  When to Use:
- You have structured numerical data  
- You know (or can guess) the number of clusters  
- Need fast, simple unsupervised learning  

### âŒ When Not to Use:
- Data isnâ€™t spherical or well-separated  
- You have outliers or varying cluster sizes  
- Need probabilistic clustering (use GMM instead)

### âœ… Example:

### ðŸ”§ Example: Low-pass filter a noisy signal
```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Sample noisy signal
t = np.linspace(0, 1, 500, False)
signal = np.sin(2 * np.pi * 5 * t) + np.random.normal(0, 0.3, t.shape)

# Design a Butterworth low-pass filter
b, a = butter(N=4, Wn=0.1, btype='low')  # Wn is frequency (0 to 1)
filtered = filtfilt(b, a, signal)

plt.plot(t, signal, label="Noisy")
plt.plot(t, filtered, label="Filtered", linewidth=2)
plt.legend(); plt.grid()
plt.title("Low-Pass Filtering with Butterworth")
plt.show()
```

ðŸ“Œ *Used in*:  
- Frequency analysis (audio, EEG, signals)  
- Signal filtering  
- Converting data from time â†’ frequency domain  

---

## ðŸ’¡ What is FFT?

**FFT (Fast Fourier Transform)** converts a **signal from time domain to frequency domain**.

- Time domain = signal over time (e.g., audio waveform)
- Frequency domain = what frequencies are present (e.g., bass, treble)

---

## ðŸ”· 1. **Core FFT Functions**

| Function     | Description                                        |
|--------------|----------------------------------------------------|
| `fft`        | Compute 1D FFT (time â†’ frequency)                  |
| `ifft`       | Inverse FFT (frequency â†’ time)                     |
| `rfft`       | FFT for **real inputs** only (faster)              |
| `irfft`      | Inverse FFT of real FFT                            |
| `fftn`       | N-Dimensional FFT (e.g., image processing)         |
| `ifftn`      | Inverse of `fftn`                                  |

---

## ðŸ”§ Basic Example: Frequency analysis of a signal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Create a signal: 5Hz + 20Hz components
t = np.linspace(0, 1, 500, False)
signal = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*20*t)

# Apply FFT
yf = fft(signal)
xf = fftfreq(len(t), 1/500)  # Frequencies

# Plot frequency domain
plt.plot(xf[:250], np.abs(yf)[:250])  # Positive frequencies
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

---

## ðŸ“˜ Function Parameters

### `fft(x, n=None, axis=-1, norm=None)`
- `x`: Input signal (1D array)
- `n`: Length of FFT (optional)
- `axis`: Axis to apply FFT on (default: last)
- `norm`: `'forward'`, `'backward'`, `'ortho'`

### `fftfreq(n, d=1.0)`
- `n`: Number of points
- `d`: Sample spacing (1/sample rate)

---

## ðŸ”„ Inverse FFT

Use `ifft` or `irfft` to convert back to time domain after processing in frequency.

```python
from scipy.fft import ifft

reconstructed_signal = ifft(yf)
```

## âŒ When NOT to Use

- When data isn't sampled uniformly (use interpolation first)

## ðŸ”· 1. **Core Mathematical Functions**

| Function     | Description                                              |

### ðŸ“˜ Example:

```python
from scipy import special

print(special.gamma(5))        # 4! = 24
print(special.loggamma(5))     # log(24)
print(special.erf(1.0))        # â‰ˆ 0.8427
print(special.expi(1.0))       # â‰ˆ 1.8951
print(special.beta(2, 3))      # â‰ˆ 0.0833
```

### ðŸ“˜ Example (Chebyshev Polynomial):
