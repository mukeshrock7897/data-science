## **1. scipy.cluster**
* The `scipy.cluster` module provides algorithms for hierarchical clustering and vector quantization, such as K-Means. Here's an in-depth explanation with use cases and examples.

### **Hierarchical Clustering**

**1. linkage**
* Performs hierarchical/agglomerative clustering.

* Use Case: Useful for creating a hierarchy of clusters where each data point starts in its own cluster, and pairs of clusters are merged as you move up the hierarchy.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
data = np.random.rand(10, 2)

# Perform hierarchical clustering
Z = linkage(data, method='ward')

# Plot dendrogram
plt.figure()
dendrogram(Z)
plt.show()
```

**2. dendrogram**
* Creates a plot of the hierarchical clustering as a dendrogram.

* Use Case: Visualize the arrangement of the clusters formed at each level of the hierarchy.

```python
from scipy.cluster.hierarchy import dendrogram

# Using the linkage matrix Z from the previous example
dendrogram(Z)
plt.show()
```

**3. inconsistent**
* Calculates inconsistency statistics for each link of the hierarchical clustering.

* Use Case: Helps in identifying the level at which clusters merge inconsistently.

```python
from scipy.cluster.hierarchy import inconsistent

# Calculate inconsistency matrix
inconsistency = inconsistent(Z)
print(inconsistency)
```

**4. cophenet**
* Calculates the cophenetic correlation coefficient to evaluate the quality of the clustering.

* Use Case: Measure how faithfully the dendrogram represents the pairwise distances between the original data points.

```python
from scipy.cluster.hierarchy import cophenet

# Calculate cophenetic correlation coefficient
result = cophenet(Z) # Change made here
c = result[0] # Change made here
coph_dists = result[1] # Change made here
print("Cophenetic correlation coefficient:", c)
```

**5. fcluster**
* Forms flat clusters from the hierarchical clustering defined by the linkage matrix.

* Use Case: Extract a specific number of flat clusters from a hierarchical cluster tree.



```python
from scipy.cluster.hierarchy import fcluster

# Form flat clusters with a max distance threshold
max_d = 0.5
clusters = fcluster(Z, max_d, criterion='distance')
print(clusters)
```

**6. fclusterdata**
* Performs hierarchical clustering and returns flat clusters directly from the data.

* Use Case: Convenience function to perform hierarchical clustering and directly get flat clusters.

```python
from scipy.cluster.hierarchy import fclusterdata

# Generate random data
data = np.random.rand(10, 2)

# Form flat clusters
clusters = fclusterdata(data, t=0.5, criterion='distance')
print(clusters)
```

**7. leaders**
* Finds the leader of each flat cluster.
* Use Case: Identify representative elements (leaders) of each cluster.



```python
from scipy.cluster.hierarchy import leaders

# Get cluster leaders
L, M = leaders(Z, clusters)
print(L, M)
```

### **Vector Quantization / K-Means**

**1. kmeans**
* Performs K-Means clustering.

* Use Case: Partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean.

```python
from scipy.cluster.vq import kmeans, vq
import numpy as np

# Generate random data
np.random.seed(42)
data = np.random.rand(100, 2)

# Compute K-Means
centroids, distortion = kmeans(data, 3)

# Assign each sample to a cluster
idx, _ = vq(data, centroids)

print("Cluster indices:", idx)
print("Centroids:", centroids)
```

**2. kmeans2**
* A variant of the K-Means algorithm with different initialization methods.

* Use Case: Provides alternative initialization methods to potentially improve convergence.

```python
from scipy.cluster.vq import kmeans2

# Compute K-Means with kmeans2
centroids, labels = kmeans2(data, 3, minit='points')
print("Cluster labels:", labels)
print("Centroids:", centroids)
```

**3. whiten**
* Normalizes a group of observations on a per-feature basis.

* Use Case: Preprocess data by scaling each feature to unit variance.

```python
from scipy.cluster.vq import whiten

# Whiten data
whitened = whiten(data)
print("Whitened data:", whitened)
```

**4. vq**
* Quantizes observation vectors to the nearest code book entries.

* Use Case: Assigns each observation to the nearest centroid from a predefined code book.



```python
# Assign each sample to a cluster using predefined centroids
idx, _ = vq(data, centroids)
print("Cluster indices:", idx)
```

## **Ordinary Differential Equations (ODE) Solvers**

**1. odeint**
* Integrate a system of ordinary differential equations.

* Use Case: Solve systems of ODEs using the LSODA method from ODEPACK.

```python
from scipy.integrate import odeint

# Define the system of ODEs
def system(y, t):
    dydt = -2 * y
    return dydt

# Solve the system
t = np.linspace(0, 5, 100)
y0 = [1]
solution = odeint(system, y0, t)
print("Solution:", solution)
```

**2. ode**
* A more general interface for integrating ODEs.

* Use Case: More control over ODE integration methods and parameters.

```python
from scipy.integrate import ode

# Define the system of ODEs
def system(t, y):
    dydt = -2 * y
    return dydt

# Create an ODE solver instance
solver = ode(system).set_integrator('dopri5')
solver.set_initial_value(1, 0)

# Solve the system
t_final = 5
dt = 0.1
while solver.successful() and solver.t < t_final:
    solver.integrate(solver.t + dt)
    print("Time:", solver.t, "Solution:", solver.y)
```

**3. solve_bvp**
* Solve boundary value problems for ODEs.

* Use Case: Solve boundary value problems for systems of ODEs.

```python
from scipy.integrate import solve_bvp

# Define the system of ODEs
def system(x, y):
    return np.vstack((y[1], -y[0]))

# Define the boundary conditions
def bc(ya, yb):
    return np.array([ya[0], yb[0] - np.exp(-1)])

# Define the initial guess for the solution
x = np.linspace(0, 1, 5)
y = np.zeros((2, x.size))
y[0] = np.exp(-x)

# Solve the boundary value problem
solution = solve_bvp(system, bc, x, y)
print("Solution:", solution.sol(x))
```

**6. griddata**
* Interpolate unstructured D-dimensional data.

* Use Case: Interpolate scattered data points in multiple dimensions.

```python
from scipy.interpolate import griddata

# Sample data
points = np.random.rand(100, 2)
values = np.sin(points[:, 0]) * np.cos(points[:, 1])
grid_x, grid_y = np.mgrid[0:1:100j, 0:1:100j]

# Perform the interpolation
grid_z = griddata(points, values, (grid_x, grid_y), method='cubic')
print("Interpolated grid values:", grid_z)
```

**7. Rbf**
* Radial basis function interpolation.

* Use Case: Interpolate scattered data points using radial basis functions.

```python
from scipy.interpolate import Rbf

# Sample data
x = np.random.rand(10, 1).flatten()
y = np.random.rand(10, 1).flatten()
z = np.sin(x) + np.cos(y)

# Create an RBF interpolator
rbf = Rbf(x, y, z, function='multiquadric')

# Use the interpolator
xi = np.linspace(0, 1, 100)
yi = np.linspace(0, 1, 100)
zi = rbf(xi, yi)
print("RBF interpolated values:", zi)
```

**8. interpn**
* Multivariate data interpolation.

* Use Case: Interpolate data points in multiple dimensions.

```python
from scipy.interpolate import interpn

# Sample data
points = (np.linspace(0, 1, 5), np.linspace(0, 1, 5))
values = np.sin(points[0][:, None]) * np.cos(points[1][None, :])
xi = (np.linspace(0, 1, 100), np.linspace(0, 1, 100))

# Perform the interpolation
zi = interpn(points, values, xi, method='linear')
print("Interpolated values:", zi)
```

**9. RectBivariateSpline**

* Minimize a scalar function of one or more variables.
* Use Case: Find the minimum of the Rosenbrock function.

```python
from scipy.optimize import minimize

# Define the Rosenbrock function
def rosen(x):
    return sum(100.0 * (x[1:] - x[:-1]**2.0)**2.0 + (1 - x[:-1])**2.0)

# Initial guess
x0 = [1.3, 0.7]

# Minimize the Rosenbrock function
result = minimize(rosen, x0)
print("Minimum value:", result.fun)
print("Optimal parameters:", result.x)
```

* Minimize a scalar function of one variable.
* Use Case: Find the minimum of a quadratic function.

```python
from scipy.optimize import minimize_scalar

# Define a quadratic function
def f(x):
    return (x - 2)**2 + 1

# Minimize the quadratic function
result = minimize_scalar(f)
print("Minimum value:", result.fun)
print("Optimal parameter:", result.x)
```

### **Multivariable Functions**

**1. minimize**
* Minimize a multivariable scalar function.
* Use Case: Optimize a multivariable function with constraints.

```python
from scipy.optimize import minimize

# Define the objective function
def objective(x):
    return x[0]**2 + x[1]**2

# Define constraints and bounds
constraints = ({'type': 'eq', 'fun': lambda x: x[0] + x[1] - 1})
bounds = [(0, None), (0, None)]

# Initial guess
x0 = [0.5, 0.5]

# Minimize the objective function with constraints
result = minimize(objective, x0, constraints=constraints, bounds=bounds)
print("Minimum value:", result.fun)
print("Optimal parameters:", result.x)
```

**2. root**
* Find a root of a multivariable function.
* Use Case: Find the roots of a system of nonlinear equations.

```python
from scipy.optimize import root

# Define the system of nonlinear equations
def system(x):
    return [x[0] + x[1] - 2, x[0] - x[1]**2]

# Initial guess
x0 = [1, 1]

# Find the roots
result = root(system, x0)
print("Roots:", result.x)
```

```python
from scipy.optimize import root_scalar

# Define the nonlinear equation
def equation(x):
    return x**3 - 2*x - 5

# Find the root
result = root_scalar(equation, bracket=[1, 3])
print("Root:", result.root)
```

### **Least-Squares**

**1. least_squares**
* Solve a nonlinear least-squares problem.
* Use Case: Fit a model to data points.



```python
from scipy.optimize import least_squares

# Define the residuals function
def residuals(x, t, y):
    return x[0] * np.exp(-x[1] * t) - y

# Data points
t = np.linspace(0, 10, 10)
y = 2.5 * np.exp(-1.3 * t) + np.random.normal(size=t.shape)

# Initial guess
x0 = [1, 1]

# Solve the least-squares problem
result = least_squares(residuals, x0, args=(t, y))
print("Optimal parameters:", result.x)
```

```python
from scipy.optimize import curve_fit

# Define the model function
def model(t, a, b):
    return a * np.exp(-b * t)

# Data points
t = np.linspace(0, 10, 10)
y = 2.5 * np.exp(-1.3 * t) + np.random.normal(size=t.shape)

# Fit the model to data points
params, covariance = curve_fit(model, t, y)
print("Optimal parameters:", params)
```

### Linear Programming

**1. linprog**
* Solve a linear programming problem.
* Use Case: Optimize a linear objective function subject to linear constraints.

```python
from scipy.optimize import linprog

# Define the objective function coefficients
c = [-1, -2]

# Define the inequality constraint matrix and vector
A = [[1, 1], [1, -1]]
b = [2, 1]

# Define the bounds for variables
x_bounds = (0, None)
y_bounds = (0, None)

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])
print("Optimal value:", result.fun)
print("Optimal parameters:", result.x)
```

### **Non-Linear Optimization**

**1. basinhopping**
* Global optimization by simulated annealing.
* Use Case: Find the global minimum of a non-convex function.

```python
from scipy.optimize import basinhopping

# Define a non-convex function
def func(x):
    return np.sin(x) + 0.1 * (x - 2)**2

# Initial guess
x0 = [1]

# Perform basinhopping
result = basinhopping(func, x0)
print("Global minimum value:", result.fun)
print("Optimal parameter:", result.x)
```

## **Global Optimizers**

**1. differential_evolution**
* Global optimization using differential evolution.
* Use Case: Find the global minimum of a function with bounds.

```python
from scipy.optimize import differential_evolution

# Define the objective function
def objective(x):
    return x[0]**2 + x[1]**2

# Define bounds
bounds = [(-5, 5), (-5, 5)]

# Perform differential evolution
result = differential_evolution(objective, bounds)
print("Global minimum value:", result.fun)
print("Optimal parameters:", result.x)
```

**2. shgo**
* Simplified Hypercube Global Optimization.
* Use Case: Perform global optimization with non-linear constraints.

```python
from scipy.optimize import shgo

# Define the objective function
def objective(x):
    return x[0]**2 + x[1]**2

# Define constraints
constraints = ({'type': 'eq', 'fun': lambda x: x[0] + x[1] - 1})

# Define bounds
bounds = [(-5, 5), (-5, 5)]

# Perform SHGO
result = shgo(objective, bounds, constraints=constraints)
print("Global minimum value:", result.fun)
print("Optimal parameters:", result.x)
```

**3. dual_annealing**
* Global optimization using dual annealing.
* Use Case: Find the global minimum of a function within given bounds.

```python
from scipy.optimize import dual_annealing

# Define the objective function
def objective(x):
    return x[0]**2 + x[1]**2

# Define bounds
bounds = [(-5, 5), (-5, 5)]

# Perform dual annealing
result = dual_annealing(objective, bounds)
print("Global minimum value:", result.fun)
print("Optimal parameters:", result.x)
```

### **scipy.stats**
* scipy.stats is a module that provides a large number of probability distributions and statistical functions.

### **Statistical Functions**

**1. Descriptive Statistics**

  * **mean:** Calculates the arithmetic mean of data.

    * Use Case: Summarize the central tendency of a dataset.

```python
import numpy as np
from scipy.stats import describe

data = np.random.normal(size=1000)
mean_value = np.mean(data)

print("Mean:", mean_value)
```

* **std:** Calculates the standard deviation of data.
  * Use Case: Measure the dispersion of a dataset.

```python
std_value = np.std(data)

print("Standard Deviation:", std_value)
```

* **var:** Calculates the variance of data.
  * Use Case: Measure the spread of a dataset.

```python
variance_value = np.var(data)

print("Variance:", variance_value)
```

### **2. Statistical Tests**
* **ttest:** Performs a T-test for the mean of one or two independent samples.

  * Use Case: Test the null hypothesis that the means of two populations are equal.

```python
from scipy.stats import ttest_ind

data1 = np.random.normal(size=100)
data2 = np.random.normal(size=100)
t_stat, p_value = ttest_ind(data1, data2)

print("T-test statistic:", t_stat)
print("P-value:", p_value)
```

* **chisquare:** Performs a chi-square test to determine if the observed frequencies match the expected frequencies.
  * Use Case: Test the independence of categorical variables.

```python
from scipy.stats import chisquare

observed = [10, 20, 30]
expected = [15, 15, 30]
chi_stat, p_value = chisquare(observed, expected)

print("Chi-square statistic:", chi_stat)
print("P-value:", p_value)
```

* **kstest:** Performs the Kolmogorov-Smirnov test for goodness of fit.
  * Use Case: Test if a sample matches a given distribution.

```python
from scipy.stats import kstest

data = np.random.normal(size=1000)
ks_stat, p_value = kstest(data, 'norm')

print("KS-test statistic:", ks_stat)
print("P-value:", p_value)
```

### **3. Correlation Functions**
* **pearsonr:** Calculates the Pearson correlation coefficient.
  * Use Case: Measure the linear relationship between two datasets.

```python
from scipy.stats import pearsonr

data1 = np.random.normal(size=100)
data2 = np.random.normal(size=100)
corr, p_value = pearsonr(data1, data2)

print("Pearson correlation:", corr)
print("P-value:", p_value)
```

* **spearmanr:** Calculates the Spearman rank-order correlation coefficient.
  * Use Case: Measure the monotonic relationship between two datasets.

```python
from scipy.stats import spearmanr

corr, p_value = spearmanr(data1, data2)

print("Spearman correlation:", corr)
print("P-value:", p_value)
```

### **Distributions**

**1. Continuous Distributions**
* **norm:** Normal (Gaussian) distribution.
  * Use Case: Model normally distributed data.

```python
from scipy.stats import norm

x = np.linspace(-5, 5, 100)
pdf = norm.pdf(x)
cdf = norm.cdf(x)

plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.title("Normal Distribution")
plt.show()
```

* **expon:** Exponential distribution.
  * Use Case: Model time between events in a Poisson process.

```python
from scipy.stats import expon

x = np.linspace(0, 5, 100)
pdf = expon.pdf(x)
cdf = expon.cdf(x)

plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.title("Exponential Distribution")
plt.show()
```

* **gamma:** Gamma distribution.
  * Use Case: Model waiting times in queuing systems.

```python
from scipy.stats import gamma

x = np.linspace(0, 10, 100)
pdf = gamma.pdf(x, a=2)
cdf = gamma.cdf(x, a=2)

plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.title("Gamma Distribution")
plt.show()
```

### **2. Discrete Distributions**
* **binom:** Binomial distribution.
  * Use Case: Model the number of successes in a fixed number of trials.

```python
from scipy.stats import binom

x = np.arange(0, 11)
pmf = binom.pmf(x, n=10, p=0.5)

plt.stem(x, pmf, basefmt=" ")
plt.title("Binomial Distribution")
plt.show()
```

* **poisson:** Poisson distribution.
  * Use Case: Model the number of events occurring in a fixed interval of time or space.

```python
from scipy.stats import poisson

x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=3)

plt.stem(x, pmf, basefmt=" ")
plt.title("Poisson Distribution")
plt.show()
```

* **geom:** Geometric distribution.
  * Use Case: Model the number of trials until the first success.

```python
from scipy.stats import geom

x = np.arange(1, 11)
pmf = geom.pmf(x, p=0.5)

plt.stem(x, pmf, basefmt=" ")
plt.title("Geometric Distribution")
plt.show()
```

### **3. Multivariate Distributions**

* **multivariate_normal:** Multivariate normal distribution.
  * Use Case: Model the joint distribution of multiple normally distributed variables.

```python
from scipy.stats import multivariate_normal

mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))
pdf = multivariate_normal.pdf(pos, mean=mean, cov=cov)

plt.contourf(x, y, pdf)
plt.title("Multivariate Normal Distribution")
plt.show()
```

* **dirichlet:** Dirichlet distribution.
  * Use Case: Model proportions that sum to one, commonly used in Bayesian statistics.

```python
from scipy.stats import dirichlet

alpha = [1, 1, 1]
samples = dirichlet.rvs(alpha, size=1000)

plt.hist(samples, bins=20, density=True, histtype='step', label=['x1', 'x2', 'x3'])
plt.title("Dirichlet Distribution")
plt.legend()
plt.show()
```

### **4. Kernel Density Estimation**
* **gaussian_kde:** Perform kernel density estimation using a Gaussian kernel.
  * **Use Case:** Estimate the probability density function of a random variable.

```python
from scipy.stats import gaussian_kde

data = np.random.normal(size=1000)
kde = gaussian_kde(data)
x = np.linspace(-5, 5, 100)
pdf = kde(x)

plt.plot(x, pdf)
plt.title("Kernel Density Estimation")
plt.show()
```
