**plt.subplot(nrows, ncols, index, **kwargs)**
* The plt.subplot() function in Matplotlib is used to create multiple subplots within a single figure.

Parameter Explanations:
* nrows → Number of rows in the subplot grid.
* ncols → Number of columns in the subplot grid.
* index → The position of the subplot in the grid (starts from 1, left to right, top to bottom).
* **kwargs → Additional keyword arguments for customization.

Optional kwargs Parameters:
* projection → Set to "polar" for a polar plot.
* sharex / sharey → Share the x/y-axis with another subplot.
* facecolor → Background color of the subplot.
* frameon → True or False (display or hide the frame).
* label → Label for the subplot (used with plt.legend()).


```python
# Example 1: Creating a 2x2 Grid of Subplots

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)  # 1st subplot
plt.plot(x, y)
plt.title("Plot 1")

plt.subplot(2, 2, 2)  # 2nd subplot
plt.plot(x, np.cos(x), 'r')
plt.title("Plot 2")

plt.subplot(2, 2, 3)  # 3rd subplot
plt.plot(x, np.tan(x), 'g')
plt.title("Plot 3")

plt.subplot(2, 2, 4)  # 4th subplot
plt.plot(x, -y, 'm')
plt.title("Plot 4")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Example 2: Sharing X and Y Axes

fig = plt.figure(figsize=(6, 4))

plt.subplot(2, 1, 1, sharex=None)  # Share X-axis
plt.plot(x, y, label="Sine Wave")
plt.legend()
plt.title("Shared X-Axis")

plt.subplot(2, 1, 2, sharex=None)  # Share X-axis
plt.plot(x, np.cos(x), label="Cosine Wave", color="r")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Example 3: Polar Plot in a Subplot


plt.subplot(1, 2, 1, projection='polar')  # First subplot as polar plot
plt.plot(x, y)
plt.title("Polar Plot")

plt.subplot(1, 2, 2)
plt.plot(x, np.sin(2*x), 'g')
plt.title("Regular Plot")

plt.show()
```
