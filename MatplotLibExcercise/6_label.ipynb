### **1. `plt.xlabel(label, **kwargs)`**  
**Adds a label to the x-axis.**  
#### **Parameters:**  
- **`label`** *(str)* → The text for the x-axis label.  
- **`fontsize`** *(int or str)* → Font size (e.g., `12`, `'large'`).  
- **`color`** *(str)* → Label color (e.g., `'red'`, `'blue'`).  
- **`fontweight`** *(str or int)* → Thickness of the text (`'bold'`, `'light'`).  
- **`rotation`** *(float or str)* → Rotates the text (`0` = horizontal, `90` = vertical).  
- **`labelpad`** *(float)* → Space between axis and label.  
- **`loc`** *(str)* → Position of the label (`'left'`, `'center'`, `'right'`).  
- **`family`** *(str)* → Font family (`'serif'`, `'sans-serif'`).  

---

### **2. `plt.ylabel(label, **kwargs)`**  
**Adds a label to the y-axis.**  
#### **Parameters:**  
(Same as `plt.xlabel()`, but applies to the y-axis.)  
- **`label`** *(str)* → The text for the y-axis label.  
- **`fontsize`**, **`color`**, **`fontweight`**, **`rotation`**, **`labelpad`**, **`loc`**, **`family`** → Work the same way as `plt.xlabel()`.  
- **Extra:** `rotation=0` makes the label horizontal instead of vertical.  

---

### **3. `plt.title(label, **kwargs)`**  
**Adds a title to the plot.**  
#### **Parameters:**  
- **`label`** *(str)* → The text for the title.  
- **`fontsize`** *(int or str)* → Font size for the title.  
- **`color`** *(str)* → Text color.  
- **`fontweight`** *(str or int)* → Text thickness (`'bold'`, `'normal'`).  
- **`loc`** *(str)* → Position of the title (`'left'`, `'center'`, `'right'`).  
- **`pad`** *(float)* → Space between the title and the plot.  
- **`backgroundcolor`** *(str)* → Background color behind the text.  

---

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

# Create plot
plt.plot(x, y, marker='o', linestyle='--', color='b', label="Data Line")

# Customize x-axis label
plt.xlabel("X-Axis Label", 
           fontsize=12, 
           color='red', 
           fontweight='bold', 
           rotation=0, 
           labelpad=10, 
           loc='center')

# Customize y-axis label
plt.ylabel("Y-Axis Label", 
           fontsize=12, 
           color='green', 
           fontweight='bold', 
           rotation=90, 
           labelpad=15, 
           loc='top')

# Customize title
plt.title("Customized Plot Title", 
          fontsize=14, 
          color='blue', 
          fontweight='bold', 
          loc='center', 
          pad=20, 
          backgroundcolor='lightgray')

# Show grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
