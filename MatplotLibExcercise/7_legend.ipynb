plt.legend(*args, **kwargs) â†’ Adds a legend to the plot, displaying labels for plotted elements.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 18]
y2 = [5, 15, 10, 20, 12]

plt.plot(x, y1, label="Line 1", color='blue')
plt.plot(x, y2, label="Line 2", color='red')

plt.legend()  # Default legend placement
plt.title("Basic Legend Example")
plt.show()
```
