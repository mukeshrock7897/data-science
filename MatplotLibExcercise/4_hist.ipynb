1. **`x`**: The input data to be histogrammed. It can be a list, NumPy array, or Pandas Series.  

2. **`bins`**: The number of bins (intervals) for grouping data. It can be:  
   - An integer (e.g., `bins=10` creates 10 equal-width bins).  
   - A sequence defining bin edges (e.g., `bins=[0, 10, 20, 30]`).  
   - `"auto"`, `"sturges"`, `"fd"`, `"doane"`, `"scott"`, `"stone"`, `"rice"`, `"sqrt"` (automatic bin selection methods).  

3. **`range`**: A tuple `(min, max)` specifying the range of data to consider.  
   - Example: `range=(0, 100)` ensures only values between 0 and 100 are binned.  
   - If `None`, the entire data range is used.  

4. **`density`**:  
   - `False` (default) → The histogram shows the count of values in each bin.  
   - `True` → The histogram is normalized to show a probability density.  

5. **`weights`**: A sequence of weights for each data point.  
   - Example: If `weights=[2, 3, 1]`, the first value contributes twice, the second three times, and so on.  

6. **`cumulative`**:  
   - `False` (default) → The histogram shows counts per bin.  
   - `True` → The histogram shows cumulative sums, meaning each bin includes all previous counts.  

### **Appearance & Styling**  
7. **`bottom`**: A scalar or array setting the starting y-position of bars (useful for stacked histograms).  

8. **`histtype`**: The style of the histogram. Options:  
   - `"bar"` (default) → Traditional vertical bars.  
   - `"barstacked"` → Stacked bars (when `stacked=True`).  
   - `"step"` → Unfilled step plot.  
   - `"stepfilled"` → Filled step plot.  

9. **`align`**: Controls bin alignment. Options:  
   - `"mid"` (default) → Bars are centered on bin edges.  
   - `"left"` → Bars start at bin edges.  
   - `"right"` → Bars end at bin edges.  

10. **`orientation`**:  
   - `"vertical"` (default) → Bars are vertical.  
   - `"horizontal"` → Bars are horizontal.  

11. **`rwidth`**: Relative width of bars as a fraction of the bin width.  
   - Example: `rwidth=0.8` means bars occupy 80% of available space.  

12. **`log`**:  
   - `False` (default) → Uses a linear scale.  
   - `True` → Uses a logarithmic scale for the y-axis.  

13. **`color`**: Defines the color of bars. Can be:  
   - A single color (`"red"`, `"#ff5733"`).  
   - A list of colors (for different bars).  
   - A colormap (`cmap="viridis"`).  

14. **`label`**: Label for the histogram (used in `legend()`).  

15. **`stacked`**:  
   - `False` (default) → Each dataset is plotted separately.  
   - `True` → Multiple datasets are stacked on top of each other.  

16. **`data`**: If using a Pandas DataFrame, this allows passing column names directly (e.g., `data=df, x="column_name"`).  

17. **`**kwargs`**: Additional customization options like `alpha` (transparency), `edgecolor`, `linewidth`, etc. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=500)  # Normal distribution (mean=50, std=10)
data2 = np.random.normal(loc=60, scale=15, size=500)  # Normal distribution (mean=60, std=15)

# Define bin edges manually
bins = np.arange(20, 100, 5)  # Bins from 20 to 100 with step of 5

# Plot histogram
plt.hist([data1, data2], bins=bins, 
         density=True,         # Show probability density instead of counts
         cumulative=False,      # Do not make it cumulative
         histtype='barstacked', # Use stacked bars
         color=['blue', 'orange'],  # Different colors for datasets
         edgecolor='black',     # Black edges for bars
         alpha=0.7,             # Transparency level
         label=['Dataset 1', 'Dataset 2'],  # Labels for legend
         rwidth=0.9,            # Bar width (90% of available space)
         log=False,             # Use linear scale
         stacked=True)          # Stacked histogram

# Add labels and title
plt.xlabel("Value Range")
plt.ylabel("Probability Density")
plt.title("Customized Histogram Example")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Add horizontal grid lines

# Show plot
plt.show()
```
