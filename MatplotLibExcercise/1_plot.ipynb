```python
#1- How do you create a simple line plot using plt.plot() with x = [1, 2, 3, 4] and y = [10, 20, 25, 30]?
import matplotlib.pyplot as plt

%matplotlib inline
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y)
plt.show()
```

```python
#2- What happens if you provide only one list to plt.plot() instead of both x and y?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x)
plt.show()
```

```python
# 3- Can x and y be NumPy arrays instead of lists?
import numpy as np


x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y)
plt.show()
```

```python
# Color Parameter Questions:
# 4- How do you change the line color to red using a color name?

import numpy as np


x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,color="red")
plt.show()
```

```python
# 5- How do you set the color to a custom hex code, e.g., #ff5733?

import numpy as np


x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,color="#ff5733")
plt.show()
```

```python
# Linestyle Parameter Questions:
# How do you create a dashed line in plt.plot()?

# Can you combine color and linestyle in a single plt.plot() call?

import matplotlib.pyplot as plt

%matplotlib inline
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, linestyle="--")
plt.title("Dashed Line Created")
plt.show()

# What is the difference between '-', '--', '-.', and ':' in linestyle?

%matplotlib inline
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, linestyle="-")
plt.title("Solid Line Created")
plt.show()

plt.plot(x,y, linestyle="--")
plt.title("Dashed Line Created")
plt.show()

plt.plot(x,y, linestyle="-.")
plt.title("Underline Line Created")
plt.show()

plt.plot(x,y, linestyle=":")
plt.title("Dotted Line Created")
plt.show()

```

```python
# Linewidth Parameter Questions:




# How do you make the line thicker using linewidth?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, linewidth=2.0)
plt.show()


# What happens if you set linewidth=0?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, linewidth=0)
plt.show()


# Can you pass a negative value to linewidth?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, linewidth=-2)
plt.show()
```

```python
# Marker Parameter Questions:

# Which marker would you use to represent data points as squares?
# How do you create a line plot where each data point is marked with a star ('*')?


# How do you add circular markers to a line plot?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, marker="o")
plt.title("Circular Marker ")
plt.show()


x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, marker="s")
plt.title("Square Marker ")
plt.show()


x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, marker="D")
plt.title("Diamond Marker")
plt.show()

x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, marker="*")
plt.title("Star Marker")
plt.show()
```

```python
# Label Parameter Questions:

# What is the purpose of the label parameter?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, label="Data Points")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()


# How do you ensure the label appears in the legend?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, label="Data Points")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Displaying the data points")
plt.legend()
plt.show()


# Can you add multiple labeled lines in the same plot?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
z = [15, 25, 30, 35]
plt.plot(x,y, label="X and Y Line Data")
plt.plot(y,z, label="Y and Z Line Data")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()

```

```python
# Alpha Parameter Questions:



# What does setting alpha=0.5 do?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, alpha=0.5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("0.5 transparent")
plt.show()


# How would you create a fully transparent line?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, alpha=0)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Fully transparent")
plt.show()


# What happens if you set alpha to a value greater than 1.0 or less than 0.0?

x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
# plt.plot(x,y, alpha=-0.5)  # alpha value can not be less than 0 or more that 1
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Putting Negative Value 0.5 transparent")
plt.show()
```

```python
# Combination Questions:

# How would you create a thick blue dashed line with circular markers?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, color="blue", marker="o" , linestyle="--", linewidth=2.0)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("thick blue dashed line with circular markers")
plt.show()


# How do you create a plot with a green solid line, square markers, and 50% transparency?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, color="green", marker="s" , linestyle="-", linewidth=2.0 , alpha=0.5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("green solid line, square markers, and 50% transparency")
plt.show()



# What happens if you use plt.plot(x, y, 'ro-')?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x,y, 'ro-')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("solid red line and circle marker short ro-")
plt.show()
```

```python
# Advanced Questions:

# Can you use different alpha values for the line and markers separately?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
z = [10, 20, 25, 30]
plt.plot(x,y, marker="o", color="pink" , linestyle="--", linewidth=2.0 , alpha=0.5 , label="0.5 transparency")
plt.plot(y,z, marker="s" ,color="cyan" , linestyle="-", linewidth=2.0 , alpha=0.8 , label="0.8 transparency")
plt.title("different-different transparency lines with different-different markers")
plt.legend()
plt.show()


# How do you plot multiple lines with different styles in the same figure?
a = [1, 2, 3, 4] 
b = [1.5, 20.5, 25.5, 30.5]
c = [10.5, 20.5, 25.5, 30.5]
d = [10.5, 20.5, 25.5, 30.5]
plt.plot(a,b, marker="o", color="pink" , linestyle="-", linewidth=2.0 , alpha=0.5 , label="solid line")
plt.plot(b,c, marker="s", color="red" , linestyle="--", linewidth=2.0 , alpha=0.5 , label="dashed line")
plt.plot(c,d, marker="D", color="blue" , linestyle="-.", linewidth=2.0 , alpha=0.5 , label="dotted dashed line")
plt.plot(c,a, marker="*", color="black" , linestyle=":", linewidth=2.0 , alpha=0.5 , label="dotted line")
plt.legend()
plt.show()


# What happens if you use plt.plot(x, y, linestyle='dotted', color='yellow', linewidth=3, marker='D')?
x = [1, 2, 3, 4] 
y = [10, 20, 25, 30]
plt.plot(x, y, linestyle='dotted', color='yellow', linewidth=3, marker='D')
plt.title("linewidth 3 ")
plt.show()
```
