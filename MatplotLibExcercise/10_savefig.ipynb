**Parameter Explanations:**
* "fname" → The filename or file path (e.g., "plot.png", "plot.pdf", "C:/path/plot.jpg").
* "dpi" → Resolution of the saved figure (default: 100, higher means better quality).
* "format" → File format ('png', 'pdf', 'svg', 'jpg', etc.).
* "bbox_inches" → Adjusts layout before saving:
        * 'tight' → Trims extra white spaces.
        * 'standard' → Keeps default padding.
* "transparent" → True makes background transparent.
* "facecolor" → Background color of the figure.
* "edgecolor" → Color of the figure border.
* "pad_inches" → Padding around the figure.
* "quality" → JPEG quality (1-100, only for JPEG format).
* "optimize" → Compress the image file (True reduces file size).
* "orientation" → 'portrait' or 'landscape' (only for PostScript formats).
* "metadata" → Dictionary of metadata (e.g., author, title).


```python
# Example 1: Save as PNG (Default)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 18]

plt.plot(x, y, marker='o')
plt.title("Line Plot Example")
plt.savefig("plot.png")  # Saves as 'plot.png' in the current directory
plt.show()
```

```python
# Example 2: Save as PDF with High Resolution
plt.plot(x, y, marker='o', color='r')
plt.title("High-Quality PDF")
plt.savefig("plot.pdf", dpi=300)  # High-resolution PDF
plt.show()
```

```python
#  Example 3: Save with Transparent Background
plt.plot(x, y, marker='o', color='g')
plt.title("Transparent Background Example")
plt.savefig("transparent_plot.png", transparent=True, dpi=200)
plt.show()
```

```python
# Example 4: Save Without Extra White Space
plt.plot(x, y, marker='o', color='b')
plt.title("Tight Layout Example")
plt.savefig("tight_plot.png", bbox_inches='tight', dpi=150)
plt.show()
```

```python
# Example 5: Save with Metadata

plt.plot(x, y, marker='o', color='m')
plt.title("Metadata Example")
plt.savefig("metadata_plot.png", metadata={'Title': 'Sample Plot', 'Author': 'Mukesh'})
plt.show()
```
