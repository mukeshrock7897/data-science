plt.scatter(x, y, s=None, c=None, *, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None, colorizer=None, plotnonfinite=False, data=None, **kwargs)
1. **`x, y`** – Arrays or lists representing the X and Y coordinates of the data points.  
2. **`s`** – Controls the size of each marker; can be a single value or an array for varying sizes.  
3. **`c`** – Defines the color of each point; accepts a single color, list, or array mapped to `cmap`.  
4. **`marker`** – Specifies the shape of markers (e.g., `'o'` for circles, `'s'` for squares).  
5. **`cmap`** – Applies a colormap when `c` is an array, mapping values to colors (e.g., `'viridis'`).  
6. **`norm`** – Normalizes color values when using colormaps, scaling them between 0 and 1.  
7. **`vmin, vmax`** – Defines the range of values for `cmap` normalization, overriding `norm`.  
8. **`alpha`** – Adjusts marker transparency, ranging from 0 (fully transparent) to 1 (opaque).  
9. **`linewidths`** – Sets the width of the marker edges (default is usually 1.0).  
10. **`edgecolors`** – Defines the edge color of markers (e.g., `'black'`, `'white'`, `'none'`).  
11. **`colorizer`** – (Rarely used) Custom function to determine marker colors dynamically.  
12. **`plotnonfinite`** – If `True`, plots NaN or infinite values instead of ignoring them.  
13. **`data`** – Allows passing a dictionary-like object to reference column names instead of arrays.  
14. **`**kwargs`** – Additional arguments like `label`, `zorder`, or `clip_on` for further customization.  


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Generate random data
np.random.seed(42)
x = np.random.rand(50) * 10  # Random X values between 0 and 10
y = np.random.rand(50) * 10  # Random Y values between 0 and 10
sizes = np.random.randint(20, 200, size=50)  # Random marker sizes
colors = np.random.rand(50)  # Random values for colormap

# Define colormap normalization with vmin & vmax inside
norm = mcolors.Normalize(vmin=0, vmax=1)

# Create scatter plot with all parameters
plt.scatter(
    x, y, 
    s=sizes,                # Marker sizes
    c=colors,               # Marker colors
    marker='D',             # Diamond-shaped markers
    cmap='plasma',          # Colormap applied to colors
    norm=norm,              # Pass vmin & vmax through norm
    alpha=0.75,             # Transparency level
    linewidths=1.5,         # Marker edge thickness
    edgecolors='black',     # Marker edge color
    plotnonfinite=True,     # Allow plotting NaN values if present
    label='Random Points'   # Legend label
)

# Customize plot
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot with All Parameters")
plt.colorbar(label="Color Scale")  # Add color scale legend
plt.legend()  # Show legend
plt.grid(True, linestyle="--", alpha=0.5)  # Add a grid

# Show plot
plt.show()
```

```python
# Normalization (norm) Questions:

# Why do we need norm in plt.scatter()?
# because it provides the flexibility to control how data values are mapped to colors when using colormaps
from matplotlib import colors

x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])
norms = colors.Normalize(vmin=clr.min(), vmax=clr.max())

plt.scatter(x,y, s=100, c=clr, cmap="plasma", norm=norms)
plt.title("scatter plot with Normalization")
plt.colorbar()
plt.show()


# How do you normalize colors between 0 and 1?
from matplotlib import colors

x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])
norms = colors.Normalize(vmin=clr.min(), vmax=clr.max())

plt.scatter(x,y, s=100, c=clr, cmap="plasma", norm=norms)
print(norms(clr))  # will show how data got normalized b/w 0 and 1 
plt.title("scatter plot with Normalization")
plt.colorbar()
plt.show()


# What is the difference between setting vmin/vmax inside Normalize() vs. passing them separately?
# The difference between setting vmin/vmax inside Normalize() versus passing them separately to plt.scatter() (or other plotting functions) comes down to how and when the normalization is applied and whether you want to reuse the same normalization for multiple plots or other purposes.
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors

data = np.array([10, 25, 50, 75])
norm = colors.Normalize(vmin=data.min(), vmax=data.max())  # vmin/vmax set here
plt.scatter(x, y, c=data, cmap='viridis', norm=norm)  # norm object used
plt.colorbar()
plt.show()

#Reusing the same normalization for other plots or calculations
normalized_data = norm(data)
print(normalized_data)


# What happens if you pass norm=None?

from matplotlib import colors

x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])

plt.scatter(x,y, s=100, c=clr, cmap="plasma", norm=None)
plt.title("scatter plot with Normalization")
plt.colorbar()
plt.show()
```

```python
# Edge Properties (linewidths & edgecolors) Questions:

# How do you add black borders around all markers?
from matplotlib import colors

x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])

plt.scatter(x,y, s=100, marker=">", edgecolors="black", c=clr, cmap="plasma")
plt.title("scatter plot black borders around all markers")
plt.colorbar()
plt.show()


# What happens if you set edgecolors='none'?
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])

plt.scatter(x,y, s=100, marker=">", edgecolors="none", c=clr, cmap="plasma")
plt.title("scatter plot none borders around all markers")
plt.colorbar()
plt.show()



# Can you vary the border width of markers?
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])

plt.scatter(x,y, s=100, marker=">", edgecolors="none",linewidths=2, c=clr, cmap="viridis")
plt.title("scatter plot vary the border width of markers")
plt.colorbar()
plt.show()


# What happens if you pass a list to linewidths?
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr=np.array([1,2,3,4])

plt.scatter(x,y, s=100, marker=">", edgecolors="none",linewidths=[1,2,3,4], c=clr, cmap="coolwarm")
plt.title("scatter plot list to linewidths markers")
plt.colorbar()
plt.show()


# Can edgecolors accept a colormap?
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
clr = np.array([1, 2, 3, 4])

# 1. Create a colormap
cmap = plt.cm.coolwarm  # Or any other colormap you prefer

# 2. Normalize your data (important!)
norm = plt.Normalize(vmin=clr.min(), vmax=clr.max())

# 3. Get the colors from the colormap using the normalized data
edge_colors = cmap(norm(clr))

# 4. Use these colors for edgecolors
plt.scatter(x, y, s=100, marker="<", edgecolors=edge_colors, linewidths=[1, 2, 3, 4], c=clr, cmap=cmap) # c and cmap are for the facecolors
plt.title("Scatter plot with colormap-based edge colors")
plt.colorbar()
plt.show()
```

```python
# Handling NaN/Infinite Values (plotnonfinite) Questions:

# What does plotnonfinite=True do?
# Nan will be skipped
x = np.array([np.nan ,1, 2, 3, 4 , np.nan])
y = np.array([np.nan ,10, 20, 25, 30, np.nan])

plt.scatter(x,y, plotnonfinite=True)
plt.show()


# What happens if you try to plot NaN values with plotnonfinite=False?
x = np.array([np.nan ,1, 2, 3, 4 , np.nan])
y = np.array([np.nan ,10, 20, 25, 30, np.nan])

plt.scatter(x,y, plotnonfinite=False)
plt.show()


x = np.array([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
y = np.array([np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])

plt.scatter(x, y, plotnonfinite=False)  # No error raised
plt.show() # Nothing will be shown


# How do you visualize missing data in a scatter plot?
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, np.nan, 4, np.nan, 6])
y = np.array([10, np.nan, 30, 40, 50, np.nan])

# Create masked arrays
mask_x = np.isnan(x)
mask_y = np.isnan(y)
mask = mask_x | mask_y # Combine masks

x_valid = x[~mask]
y_valid = y[~mask]

x_missing = x[mask]
y_missing = y[mask]


plt.scatter(x_valid, y_valid, label="Valid Data")  # Plot valid data as usual
plt.scatter(x_missing, y_missing, marker='x', s=100, c='red', label="Missing Data")  # Mark missing data with a special marker

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot with Missing Data Visualization")
plt.legend()
plt.show()
```

```python
# Combination & Customization Questions:

# How do you create a scatter plot where marker size and color vary based on values?
x = np.array([1, 2, 3, 4 ])
y = np.array([10, 20, 25, 30])
sizes = np.array([50, 100, 150, 200])  # Varying sizes
colors = np.array([1, 2, 3, 4])  # Numerical values for colormap

plt.scatter(x,y, c=colors , s=sizes)
plt.show()


# How do you add a color bar that reflects marker colors?
x = np.array([1, 2, 3, 4 ])
y = np.array([10, 20, 25, 30])
sizes = np.array([50, 100, 150, 200])  # Varying sizes
colors = np.array([1, 2, 3, 4])  # Numerical values for colormap

scatter = plt.scatter(x, y, c=colors, s=sizes, cmap='viridis')  # Store the return value of scatter
cbar = plt.colorbar(scatter)       # Pass the scatter plot object to colorbar
cbar.set_label("Color Value") 
plt.show()


# How do you label individual points in a scatter plot?
x = np.array([1, 2, 3, 4 ])
y = np.array([10, 20, 25, 30])
colors = np.array([1, 2, 3, 4])  
labels = ['A', 'B', 'C', 'D']  # Labels for each point
plt.scatter(x, y, c=colors, cmap='viridis')

# Add labels to points
for i in range(len(x)):
    plt.text(x[i], y[i], labels[i], fontsize=12, ha='right', va='bottom', color='black')
plt.show()



# What is the effect of setting c and edgecolors to different colormaps?
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
colors = np.array([1, 2, 3, 4])  

# Normalize colors for edge mapping
normalized_colors = colors / max(colors)  # Normalize values between 0 and 1

plt.scatter(
    x, y, 
    c=colors, cmap='viridis',  # Face color uses 'viridis'
    edgecolors=plt.cm.coolwarm(normalized_colors),  # Edge color uses 'coolwarm'
    linewidths=2
)

plt.title("Effect of Setting c and edgecolors to Different Colormaps")
plt.colorbar(label="Face Color Mapping")
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()



# How do you create a scatter plot with both filled and hollow markers?

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([10, 20, 25, 30, 15, 22])
colors = np.array([1, 2, 3, 4, 5, 6])  

# Define filled and hollow markers
filled_markers = ['o', 's']  # Circle and Square (Filled)
hollow_markers = ['^', 'D']  # Triangle and Diamond (Hollow)

# Scatter for filled markers
plt.scatter(x[:3], y[:3], c=colors[:3], cmap='viridis', marker=filled_markers[1], edgecolors='black', linewidths=1.5)

# Scatter for hollow markers (facecolor set to 'none')
plt.scatter(x[3:], y[3:], c=colors[3:], cmap='coolwarm', marker=hollow_markers[1], edgecolors='black', facecolors='none', linewidths=2)

# Formatting
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot with Filled and Hollow Markers")
plt.colorbar(label="Color Scale")
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```
