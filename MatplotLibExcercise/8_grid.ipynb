plt.grid(visible=None, which='major', axis='both', **kwargs)
The plt.grid() function controls the visibility and appearance of the grid in a plot.

Parameter Explanations:
* visible → True (show grid), False (hide grid), None (toggle default behavior).
* which → 'major' (default, shows major grid lines), 'minor' (shows minor grid lines), 'both' (shows both).
* axis → 'both' (default, grid for both x and y), 'x' (grid for x-axis only), 'y' (grid for y-axis only).
* **kwargs → Additional styling options like color, linestyle, linewidth, and alpha

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 18]

plt.plot(x, y, marker='o')
plt.grid(True)  # Enable grid
plt.title("Basic Grid Example")
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.grid(True, axis='y')  # Grid for Y-axis only
plt.title("Grid on Y-Axis Only")
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)  # Dashed, light grid
plt.title("Custom Grid Style")
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.minorticks_on()  # Enable minor ticks
plt.grid(True, which='major', linestyle='-', linewidth=0.8)  # Major grid
plt.grid(True, which='minor', linestyle=':', linewidth=0.5, color='gray')  # Minor grid

plt.title("Major and Minor Grid Example")
plt.show()
```
