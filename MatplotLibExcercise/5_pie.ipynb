### **Main Parameters**  
1. **`x`**: Values for each slice (e.g., sales data, category counts).  

2. **`explode`**: List of fractions to "explode" (pull out) slices.  
   - Example: `explode=[0, 0.1, 0, 0.2]` pulls the 2nd and 4th slices.  

3. **`labels`**: Labels for each slice.  

4. **`colors`**: Custom colors for slices (e.g., `['red', 'blue', 'green']`).  

5. **`autopct`**: Format string for displaying percentage values inside slices.  
   - Example: `autopct='%1.1f%%'` shows one decimal place.  

6. **`pctdistance`**: Distance of percentage labels from the center (default `0.6`).  
   - Example: `pctdistance=0.8` moves them outward.  

7. **`shadow`**: Adds a shadow effect (`True` or `False`).  

8. **`labeldistance`**: Distance of labels from center (default `1.1`).  
   - Example: `labeldistance=1.3` moves labels further out.  

9. **`startangle`**: Rotates the pie chart (degrees, default `0`).  
   - Example: `startangle=90` starts from the top.  

10. **`radius`**: Size of the pie chart (default `1`).  

11. **`counterclock`**: Draw slices counterclockwise (`True`, default) or clockwise (`False`).  

12. **`wedgeprops`**: Dictionary for customizing wedge appearance (e.g., edge color, width).  
   - Example: `wedgeprops={'edgecolor': 'black', 'linewidth': 2}`.  

13. **`textprops`**: Dictionary for styling text inside slices.  
   - Example: `textprops={'fontsize': 12, 'color': 'black'}`.  

14. **`center`**: (x, y) position of the pie chart (default `(0,0)`).  

15. **`frame`**: Draws a rectangle around the pie (`True` or `False`).  

16. **`rotatelabels`**: Rotates slice labels to align with their sections (`True` or `False`).  

17. **`normalize`**: Normalizes values to sum to 1 (`True`, default) or keeps raw values (`False`).  

18. **`hatch`**: Adds hatching patterns (`'/', '\\', '|', '-', '+', 'x', 'o'`).  

19. **`data`**: If using a Pandas DataFrame, allows referencing column names directly.  


```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [30, 20, 25, 15, 10]  # Percentage distribution
labels = ['A', 'B', 'C', 'D', 'E']  # Category labels
colors = ['red', 'blue', 'green', 'purple', 'orange']  # Custom colors
explode = [0, 0.1, 0, 0.2, 0]  # Explode B and D slices

# Create Pie Chart
plt.pie(sizes, 
        explode=explode,  
        labels=labels,  
        colors=colors,  
        autopct='%1.1f%%',  # Show percentages
        pctdistance=0.7,  # Adjust percentage label position
        shadow=True,  # Add shadow effect
        labeldistance=1.2,  # Move labels outward
        startangle=140,  # Rotate starting position
        radius=1,  # Control overall size
        counterclock=False,  # Draw clockwise
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},  # Edge styling
        textprops={'fontsize': 12, 'color': 'black'},  # Customize text
        center=(0, 0),  # Position the pie chart
        frame=True,  # Add a border frame
        rotatelabels=True,  # Rotate labels along slices
        normalize=True,  # Normalize values to sum to 1
        hatch=['/', '\\', '|', '-', '+'])  # Different patterns for slices

# Add a title
plt.title("Customized Pie Chart Example", fontsize=14, fontweight='bold')

# Show the chart
plt.show()
```
