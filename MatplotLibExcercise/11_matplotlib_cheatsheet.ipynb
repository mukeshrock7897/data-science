# **📊Matplotlib Cheatsheet**

---

### 🧱 1. Basic Setup  
📌 Import + basic plot setup  
```python
import matplotlib.pyplot as plt  # 📥

plt.plot(x, y)          # 📈 Line plot
plt.show()              # 👁 Show the plot
```

---

### 📌 2. Plot Types  
📌 Most-used plot types  
```python
plt.plot(x, y)           # 📈 Line plot  
plt.scatter(x, y)        # 🔵 Scatter plot  
plt.bar(x, y)            # 📊 Bar chart  
plt.hist(data)           # 🟦 Histogram  
plt.boxplot(data)        # 📦 Box plot  
plt.pie(sizes)           # 🥧 Pie chart  
plt.fill_between(x, y1, y2)  # 🌊 Area plot  
plt.errorbar(x, y, yerr=errors)  # ⚠️ Error bars  
```

---

### 🎨 3. Customization  
📌 Customize labels, titles, and visuals  
```python
plt.title("Title")                    # 🏷 Title  
plt.xlabel("X Axis")                  # ⬅️ X-axis label  
plt.ylabel("Y Axis")                  # ⬆️ Y-axis label  
plt.legend(["Line A"])                # 🧾 Legend  
plt.grid(True)                        # 🧱 Grid lines  
plt.xticks(rotation=45)              # 🔄 Rotate labels  
```

---

### 🎯 4. Figure & Subplots  
📌 Multiple plots in one figure  
```python
fig, ax = plt.subplots()                      # 📐 Create one plot  
fig, axs = plt.subplots(2, 2)                 # 🧩 Grid of subplots  
axs[0, 0].plot(x, y)                          # 🔄 Use individual axis  
fig.suptitle("Main Title")                   # 🧵 Global title  
```

---

### 🖌 5. Style & Themes  
📌 Change plot styles easily  
```python
plt.style.use('ggplot')            # 🎨 Use ggplot style  
plt.style.available                 # 📚 List styles  
```

---

### 📏 6. Size, Limits & Ticks  
📌 Adjust layout and axes  
```python
plt.figure(figsize=(10, 5))         # 📐 Set figure size  
plt.xlim(0, 100)                    # ➡️ Set X-axis range  
plt.ylim(0, 50)                     # ⬆️ Set Y-axis range  
plt.xticks([0, 10, 20])             # 🔢 Custom ticks  
```

---

### 💾 7. Save Plot  
📌 Save plot as image  
```python
plt.savefig("plot.png", dpi=300)    # 💾 Save high-res  
```

---

### 🎯 8. Annotations  
📌 Highlight specific points  
```python
plt.annotate('Important', xy=(x, y), xytext=(x+1, y+5),
             arrowprops=dict(facecolor='black'))  # 📌 Add annotation  
```

---

### ⚙️ 9. Plot in Pandas  
📌 Directly plot from DataFrame  
```python
df.plot()                        # 🧠 Line plot  
df.plot(kind='bar')             # 📊 Bar chart  
df.plot(kind='box')             # 📦 Box plot  
df['col'].plot.hist()           # 🟫 Histogram  
```

---

### 🧪 10. Extras & Tips  
📌 Useful tricks  
```python
plt.tight_layout()              # 🤏 Avoid label cut-off  
plt.clf()                       # 🧼 Clear current figure  
plt.close()                     # ❌ Close figure  
```

### 🧠 6. Actual vs Predicted (Regression)  
📌 See how close predictions are to real values  
```python
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # 📈 Identity line
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.grid(True)
plt.show()
```

---

### 📦 7. Residual Plot (Regression Errors)  
📌 Diagnose prediction errors  
```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.grid(True)
plt.show()
```

### 🔍 8. Class Distribution (Target Analysis)  
📌 Check class imbalance before training
