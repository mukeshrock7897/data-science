```python
import pandas as pd
import numpy as np

# Sample data with missing values
data = {'Name': ['Alice', 'Bob', None],
        'Age': [25, None, 35],
        'Salary': [70000, 80000, None]}
df = pd.DataFrame(data)

# 1. Handling Missing Data
# 1.1. Identifying Missing Data (isna(), notna())
print("\n1.1. Identifying Missing Data (isna()):\n", df.isna())
print("\n1.1. Identifying Missing Data (notna()):\n", df.notna())

# 1.2. Dropping Missing Data (dropna())
df_dropped = df.dropna()
print("\n1.2. Dropping Rows with Missing Data:\n", df_dropped)

df_dropped_col = df.dropna(axis=1)
print("\n1.2. Dropping Columns with Missing Data:\n", df_dropped_col)

# 1.3. Filling Missing Data (fillna())
df_filled = df.fillna(0)
print("\n1.3. Filling Missing Data with 0:\n", df_filled)

df_ffill = df.ffill()  # Use ffill() directly
print("\n1.3. Forward Filling Missing Data:\n", df_ffill)

# 1.4. Interpolating Missing Data (interpolate())
df_interpolated = df.infer_objects(copy=False).interpolate()  # infer_objects first
print("\n1.4. Interpolating Missing Data:\n", df_interpolated)

# 2. Handling Duplicates (drop_duplicates())
data_dup = {'Name': ['Alice', 'Bob', 'Bob'],
            'Age': [25, 30, 30],
            'Salary': [70000, 80000, 80000]}
df_dup = pd.DataFrame(data_dup)
df_unique = df_dup.drop_duplicates()
print("\n2. Dropping Duplicate Rows:\n", df_unique)

# 3. Data Transformation
# 3.1. Replacing Values (replace())
df_replaced = df.replace(70000, 75000)
print("\n3.1. Replacing Specific Values:\n", df_replaced)

# 3.2. Renaming Columns (rename())
df_renamed = df.rename(columns={'Salary': 'Income'})
print("\n3.2. Renaming a Column:\n", df_renamed)

# 4. Changing Data Types (astype())
df_converted = df.astype({'Age': 'Int64'})  # Use 'Int64' for nullable integers
print("\n4. Changing the Data Type of a Column:\n", df_converted)

# 5. Binning and Categorizing Data (cut(), qcut())
bins = [0, 18, 35, 60]
labels = ['Teen', 'Adult', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
print("\n5. Binning Data into Intervals:\n", df)

# 6. Scaling and Normalizing Data
df['Salary_Scaled'] = df['Salary'] / 1000
print("\n6. Scaling Salary by Dividing by 1000:\n", df)

# 7. String Operations
df['Name_Lower'] = df['Name'].str.lower()
print("\n7. Converting 'Name' column to lowercase:\n", df)

# 8. Outlier Detection and Removal
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
df_outliers = df[~((df['Salary'] < (Q1 - 1.5 * IQR)) | (df['Salary'] > (Q3 + 1.5 * IQR)))]
print("\n8. Detecting Outliers Using IQR:\n", df_outliers)
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame
data = {'A': [1, 2, np.nan], 'B': [4, np.nan, 6]}
df = pd.DataFrame(data)

# 1. Identifying Missing Data
# 1.1. isnull() and notnull()
missing_data = df.isnull()
print("\n1.1. Identifying missing data (isnull()):\n", missing_data)

not_missing_data = df.notnull()
print("\n1.1. Identifying missing data (notnull()):\n", not_missing_data)

# 2. Dropping Missing Data
# 2.1. dropna() with axis and how
df_dropped_rows = df.dropna(axis=0, how='any')
print("\n2.1. Dropping rows with any missing values:\n", df_dropped_rows)

df_dropped_columns = df.dropna(axis=1, how='any')
print("\n2.1. Dropping columns with any missing values:\n", df_dropped_columns)

# 3. Filling Missing Data
# 3.1. fillna() with value, method
df_filled_value = df.fillna(0)
print("\n3.1. Filling missing values with a specific value:\n", df_filled_value)

df_filled_ffill = df.ffill()  # Use ffill() directly (FutureWarning addressed)
print("\n3.1. Forward filling missing values:\n", df_filled_ffill)

df_filled_bfill = df.bfill()  # Use bfill() directly (FutureWarning addressed)
print("\n3.1. Backward filling missing values:\n", df_filled_bfill)

# 4. Interpolating Missing Data
# 4.1. interpolate()
df_interpolated = df.infer_objects(copy=False).interpolate()  # infer_objects first (FutureWarning addressed)
print("\n4.1. Interpolating missing values:\n", df_interpolated)

# 5. Replacing Missing Data
# 5.1. replace()
df_replaced = df.replace(np.nan, -1)
print("\n5.1. Replacing NaN with a specific value:\n", df_replaced)

# 6. Checking for Missing Data
# 6.1. isnull().sum()
missing_count = df.isnull().sum()
print("\n6.1. Checking for missing data count:\n", missing_count)

# 7. Advanced Techniques
# 7.1. pd.Series with isna()
series_missing = pd.Series([1, np.nan, 3]).isna()
print("\n7.1. pd.Series with isna():\n", series_missing)

# 7.2. Custom Functions (Example)
def fill_missing(column):
    if column.isnull().any():
        return column.fillna(column.mean()) # Fill with mean if missing values
    return column

df_custom_filled = df.apply(fill_missing)
print("\n7.2. Custom Function for filling missing values:\n", df_custom_filled)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 30, 22, 35],
    'Salary': [50000, 60000, 45000, 70000]
}
df = pd.DataFrame(data)

# 1. Filtering Data
# 1.1. Boolean indexing
filtered_age = df[df['Age'] > 25]
print("\n1.1. Filtering using boolean indexing:\n", filtered_age)

# 1.2. query()
filtered_query = df.query('Salary > 55000')
print("\n1.2. Filtering using query():\n", filtered_query)

# 2. Filtering with Multiple Conditions
# 2.1. & (and), | (or)
filtered_multiple = df[(df['Age'] > 25) & (df['Salary'] > 55000)]
print("\n2.1. Filtering with multiple conditions:\n", filtered_multiple)

# 3. Sorting Data
# 3.1. sort_values()
sorted_by_age = df.sort_values(by='Age')
print("\n3.1. Sorting by a single column:\n", sorted_by_age)

sorted_by_multiple = df.sort_values(by=['Salary', 'Age'], ascending=[True, False])
print("\n3.1. Sorting by multiple columns:\n", sorted_by_multiple)

# 4. Sorting by Index
# 4.1. sort_index()
df_indexed = df.set_index('Name')
sorted_by_index = df_indexed.sort_index()
print("\n4.1. Sorting by index:\n", sorted_by_index)

# 5. Sorting with NaN Values
# 5.1. na_position
data_nan = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, None, 22, 35],
    'Salary': [50000, 60000, None, 70000]
}
df_nan = pd.DataFrame(data_nan)

sorted_nan = df_nan.sort_values(by='Age', na_position='last')
print("\n5.1. Sorting with NaN values at the end:\n", sorted_nan)

# 6. Resetting Index After Sorting
# 6.1. reset_index()
sorted_reset_index = df.sort_values(by='Age').reset_index(drop=True)
print("\n6.1. Resetting index after sorting:\n", sorted_reset_index)
```

```python
import pandas as pd

# 1. Merging DataFrames
# 1.1. merge()
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Salary': [50000, 60000, 70000]})

merged_inner = pd.merge(df1, df2, on='EmployeeID', how='inner')
print("\n1.1. Merging DataFrames (inner):\n", merged_inner)

merged_outer = pd.merge(df1, df2, on='EmployeeID', how='outer')
print("\n1.1. Merging DataFrames (outer):\n", merged_outer)


# 2. Joining DataFrames
# 2.1. join()
df3 = pd.DataFrame({
    'Salary': [50000, 60000, 70000]},
    index=[1, 2, 3])
df4 = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance']},
    index=[1, 2, 4])

joined_inner = df3.join(df4, how='inner')
print("\n2.1. Joining DataFrames (inner):\n", joined_inner)

joined_outer = df3.join(df4, how='outer')
print("\n2.1. Joining DataFrames (outer):\n", joined_outer)


# 3. Concatenating DataFrames
# 3.1. concat()
df5 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [24, 30]})
df6 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Age': [22, 35]})

concatenated_rows = pd.concat([df5, df6], axis=0, ignore_index=True)
print("\n3.1. Concatenating DataFrames vertically:\n", concatenated_rows)

df7 = pd.DataFrame({
    'Salary': [50000, 60000]})
concatenated_columns = pd.concat([df5, df7], axis=1)
print("\n3.1. Concatenating DataFrames horizontally:\n", concatenated_columns)


# 4. Handling Duplicates in Concatenation
# 4.1. drop_duplicates()
df8 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [24, 30]})
df9 = pd.DataFrame({
    'Name': ['Alice', 'David'],
    'Age': [24, 35]})

concatenated_unique = pd.concat([df8, df9]).drop_duplicates().reset_index(drop=True)
print("\n4.1. Concatenating and removing duplicates:\n", concatenated_unique)


# 5. Concatenating with Different Columns
# 5.1. concat() with different columns
df10 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [24, 30]})
df11 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Salary': [45000, 70000]})

concatenated_diff_columns = pd.concat([df10, df11], axis=0, ignore_index=True)
print("\n5.1. Concatenating with different columns:\n", concatenated_diff_columns)
```

```python
import pandas as pd

# 1. Introduction to GroupBy
# 1.1. groupby()
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Values': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)

grouped = df.groupby('Category')
print("\n1.1. Grouping by 'Category':\n", grouped)

# 2. Aggregating Data
# 2.1. sum(), mean()
aggregated_sum = grouped.sum()
print("\n2.1. Aggregating data using sum:\n", aggregated_sum)

aggregated_mean = grouped.mean()
print("\n2.1. Aggregating data using mean:\n", aggregated_mean)

# 3. Applying Multiple Aggregation Functions
# 3.1. agg() (Corrected)
aggregated_multiple = grouped.agg({'Values': ['sum', 'mean', 'count']})  # Dictionary for correct aggregation
print("\n3.1. Applying multiple aggregation functions:\n", aggregated_multiple)


# 4. Grouping by Multiple Columns
# 4.1. groupby() with multiple columns
data_multi = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Subcategory': ['X', 'Y', 'Y', 'X', 'X', 'Y'],
    'Values': [10, 20, 30, 40, 50, 60]
}
df_multi = pd.DataFrame(data_multi)

grouped_multi = df_multi.groupby(['Category', 'Subcategory']).sum()
print("\n4.1. Grouping by 'Category' and 'Subcategory':\n", grouped_multi)

# 5. Filtering Groups
# 5.1. filter()
filtered_groups = grouped.filter(lambda x: x['Values'].sum() > 50)
print("\n5.1. Filtering groups where the sum of values is greater than 50:\n", filtered_groups)

# 6. Transforming Data
# 6.1. transform()
transformed = grouped.transform('mean')
print("\n6.1. Transforming data to get the mean of each group:\n", transformed)

# 7. Custom Aggregation Functions
# 7.1. agg() with custom function
def range_func(x):
    return x.max() - x.min()

custom_agg = grouped.agg({'Values': range_func}) # Dictionary for custom function
print("\n7.1. Applying custom aggregation function:\n", custom_agg)

# 8. GroupBy with Time Series Data
# 8.1. Grouper
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
df_time = pd.DataFrame(date_rng, columns=['date'])
df_time['data'] = pd.Series(range(1, len(df_time) + 1))

df_time.set_index('date', inplace=True)
grouped_time = df_time.groupby(pd.Grouper(freq='2D')).sum()
print("\n8.1. Grouping by day (2D frequency):\n", grouped_time)
```

```python
import pandas as pd

# 1. Introduction to Aggregation
# 1.1. Aggregation functions (sum, mean, count, min, max)
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Values': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)

aggregated_sum = df.groupby('Category')['Values'].sum()
print("\n1.1. Aggregating data using sum:\n", aggregated_sum)

# 2. Descriptive Statistics
# 2.1. describe()
descriptive_stats = df['Values'].describe()
print("\n2.1. Generating descriptive statistics:\n", descriptive_stats)

# 3. Applying Multiple Aggregation Functions
# 3.1. agg()
aggregated_multiple = df.groupby('Category')['Values'].agg(['sum', 'mean', 'count'])
print("\n3.1. Applying multiple aggregation functions:\n", aggregated_multiple)

# 4. Custom Aggregation Functions
# 4.1. agg() with custom function
def range_func(x):
    return x.max() - x.min()

custom_agg = df.groupby('Category')['Values'].agg(range=range_func)
print("\n4.1. Applying custom aggregation function:\n", custom_agg)

# 5. Descriptive Statistics for Categorical Data
# 5.1. value_counts()
data_cat = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Subcategory': ['X', 'Y', 'Y', 'X', 'X', 'Y']
}
df_cat = pd.DataFrame(data_cat)

category_counts = df_cat['Category'].value_counts()
print("\n5.1. Descriptive statistics for categorical data:\n", category_counts)

# 6. Correlation and Covariance
# 6.1. corr(), cov()
data_corr = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df_corr = pd.DataFrame(data_corr)

correlation = df_corr.corr()
print("\n6.1. Calculating correlation:\n", correlation)

covariance = df_corr.cov()
print("\n6.1. Calculating covariance:\n", covariance)

# 7. Grouping and Aggregating with Descriptive Statistics
# 7.1. groupby() followed by describe()
grouped_descriptive = df.groupby('Category')['Values'].describe()
print("\n7.1. Grouping and generating descriptive statistics:\n", grouped_descriptive)
```

```python
import pandas as pd

# 1. Melt
# 1.1. melt()
data = {
    'Date': ['2023-01-01', '2023-01-02'],
    'Sales_A': [100, 150],
    'Sales_B': [200, 250]
}
df = pd.DataFrame(data)

melted_df = pd.melt(df, id_vars=['Date'], value_vars=['Sales_A', 'Sales_B'],
                    var_name='Category', value_name='Sales')
print("\n1.1. Melting the DataFrame:\n", melted_df)

# 2. Stack
# 2.1. stack()
data_stack = {
    'Date': ['2023-01-01', '2023-01-02'],
    'Sales_A': [100, 150],
    'Sales_B': [200, 250]
}
df_stack = pd.DataFrame(data_stack).set_index('Date')

stacked_df = df_stack.stack()
print("\n2.1. Stacking the DataFrame:\n", stacked_df)

# 3. Unstack
# 3.1. unstack()
unstacked_df = stacked_df.unstack()
print("\n3.1. Unstacking the stacked DataFrame:\n", unstacked_df)

# 4. Reshaping with MultiIndex
# 4.1. stack(), unstack() with MultiIndex
data_multi = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Sales': [100, 200, 150, 250]
}
df_multi = pd.DataFrame(data_multi)

df_multi.set_index(['Date', 'Category'], inplace=True)

stacked_multi = df_multi.stack()
print("\n4.1. Stacking with MultiIndex:\n", stacked_multi)

unstacked_multi = stacked_multi.unstack()
print("\n4.1. Unstacking with MultiIndex:\n", unstacked_multi)

# 5. Using pivot for Reshaping
# 5.1. pivot()
data_pivot = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Sales': [100, 200, 150, 250]
}
df_pivot = pd.DataFrame(data_pivot)

pivoted_df = df_pivot.pivot(index='Date', columns='Category', values='Sales')
print("\n5.1. Pivoting the DataFrame:\n", pivoted_df)
```

```python
import pandas as pd

# 1. Creating a DataFrame with Text Data
# 1.1. Dictionary or list of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Chef'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print("\n1.1. Creating a DataFrame with text data:\n", df)
# Expected Output:
#       Name Occupation         City
# 0    Alice   Engineer     New York
# 1      Bob     Doctor  Los Angeles
# 2  Charlie     Artist      Chicago
# 3    David       Chef      Houston

# 2. String Methods in Pandas
# 2.1. .str accessor
df['Occupation'] = df['Occupation'].str.lower()
print("\n2.1. Converting text to lowercase:\n", df)
# Expected Output:
#       Name Occupation         City
# 0    Alice   engineer     New York
# 1      Bob     doctor  Los Angeles
# 2  Charlie     artist      Chicago
# 3    David       chef      Houston

df['Is_Engineer'] = df['Occupation'].str.contains('engineer')
print("\n2.1. Checking if a string contains a substring:\n", df)
# Expected Output:
#       Name Occupation         City  Is_Engineer
# 0    Alice   engineer     New York         True
# 1      Bob     doctor  Los Angeles        False
# 2  Charlie     artist      Chicago        False
# 3    David       chef      Houston        False

# 3. String Manipulation
# 3.1. str.split(), str.join(), str.replace(), str.strip()
df['City_Split'] = df['City'].str.split(' ')
print("\n3.1. Splitting a string into a list:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split  Is_Engineer
# 0    Alice   engineer     New York         [New, York]         True
# 1      Bob     doctor  Los Angeles     [Los, Angeles]        False
# 2  Charlie     artist      Chicago          [Chicago]        False
# 3    David       chef      Houston          [Houston]        False

df['City_Joined'] = df['City_Split'].str.join(', ')
print("\n3.1. Joining a list into a string:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer
# 0    Alice   engineer     New York         [New, York]              New, York         True
# 1      Bob     doctor  Los Angeles     [Los, Angeles]        Los, Angeles        False
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False
# 3    David       chef      Houston          [Houston]              Houston        False

df['City'] = df['City'].str.replace('New York', 'NYC')
print("\n3.1. Replacing substrings:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer
# 0    Alice   engineer          NYC         [New, York]              New, York         True
# 1      Bob     doctor  Los Angeles     [Los, Angeles]        Los, Angeles        False
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False
# 3    David       chef      Houston          [Houston]              Houston        False

df['City'] = df['City'].str.strip()
print("\n3.1. Stripping whitespace:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer
# 0    Alice   engineer          NYC         [NYC]              NYC         True
# 1      Bob     doctor  Los Angeles     [Los, Angeles]        Los, Angeles        False
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False
# 3    David       chef      Houston          [Houston]              Houston        False


# 4. Text Data Analysis
# 4.1. String methods for analysis
df['Occupation_Length'] = df['Occupation'].str.len()
print("\n4.1. Counting the number of characters:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer  Occupation_Length
# 0    Alice   engineer          NYC         [NYC]              NYC         True                  7
# 1      Bob     doctor  Los Angeles     [Los, Angeles]        Los, Angeles        False                  6
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False                  6
# 3    David       chef      Houston          [Houston]              Houston        False                  4

df['A_Count'] = df['Name'].str.count('a')
print("\n4.1. Counting occurrences of a specific character:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer  Occupation_Length  A_Count
# 0    Alice   engineer          NYC         [NYC]              NYC         True                  7        1
# 1      Bob     doctor  Los Angeles     [Los, Angeles]        Los, Angeles        False                  6        0
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False                  6        1
# 3    David       chef      Houston          [Houston]              Houston        False                  4        1

# 5. Handling Missing Values in Text Data
# 5.1. fillna()
df.loc[1, 'Occupation'] = None
print("\n5.1. Introducing missing values:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer  Occupation_Length  A_Count
# 0    Alice   engineer          NYC         [NYC]              NYC         True                  7        1
# 1      Bob      None  Los Angeles     [Los, Angeles]        Los, Angeles        False                  6        0
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False                  6        1
# 3    David       chef      Houston          [Houston]              Houston        False                  4        1

df['Occupation'] = df['Occupation'].fillna('Unknown')
print("\n5.1. Filling missing values:\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer  Occupation_Length  A_Count
# 0    Alice   engineer          NYC         [NYC]              NYC         True                  7        1
# 1      Bob    Unknown  Los Angeles     [Los, Angeles]        Los, Angeles        False                  6        0
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False                  6        1
# 3    David       chef      Houston          [Houston]              Houston        False                  4        1

# 6. Regular Expressions with Text Data
# 6.1. str.contains(), str.match(), str.replace()
df['Contains_A'] = df['Occupation'].str.contains('a', case=False)
print("\n6.1. Checking if the occupation contains 'a':\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer  Occupation_Length  A_Count  Contains_A
# 0    Alice   engineer          NYC         [NYC]              NYC         True                  7        1         True
# 1      Bob    Unknown  Los Angeles     [Los, Angeles]        Los, Angeles        False                  6        0        False
# 2  Charlie     artist      Chicago          [Chicago]              Chicago        False                  6        1         True
# 3    David       chef      Houston          [Houston]              Houston        False                  4        1        False

df['Occupation'] = df['Occupation'].str.replace(r'.*a.*', 'Artist', regex=True)
print("\n6.1. Replacing occupations containing 'a' with 'Artist':\n", df)
# Expected Output:
#       Name Occupation         City       City_Split          City_Joined  Is_Engineer  Occupation_Length  A_Count  Contains_A
# 0    Alice     Artist          NYC         [NYC]              NYC         True                  7        1         True
# 1      Bob     Artist  Los Angeles     [Los, Angeles]        Los, Angeles        False                  6        0        False
# 2  Charlie     Artist      Chicago          [Chicago]              Chicago        False                  6        1         True
# 3    David       chef      Houston          [Houston]              Houston        False                  4        1        False


# 7. Concatenating and Joining Text Data
# 7.1. str.cat()  (Using + for concatenation, as str.cat is less common)
df['Name_Occupation'] = df['Name'] + ' is a ' + df['Occupation']
print(df)
```
