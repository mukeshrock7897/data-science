### DateTime

```python
import numpy as np

# 1. Basics of Date and Time

date = np.datetime64('2024-03-15')  # Date
datetime = np.datetime64('2024-03-15T10:30:00')  # Date and time
print("Date:\n", date)
# Output: Date:
# 2024-03-15
print("\n\n")

print("Datetime:\n", datetime)
# Output: Datetime:
# 2024-03-15T10:30:00
print("\n\n")

date_year = np.datetime64('2024', 'Y')
print("Date with year precision:\n", date_year)
# Output: Date with year precision:
# 2024
print("\n\n")

datetime_ns = np.datetime64('2024-03-15T10:30:00.123456789', 'ns')
print("Datetime with nanosecond precision:\n", datetime_ns)
# Output: Datetime with nanosecond precision:
# 2024-03-15T10:30:00.123456789
print("\n\n")

# 2. Representing Date & Time

dates = np.array(['2024-03-10', '2024-03-15', '2024-03-20'], dtype='datetime64')
print("Array of datetimes:\n", dates)
# Output: Array of datetimes:
# ['2024-03-10' '2024-03-15' '2024-03-20']
print("\n\n")

# 3. Date & Time Arithmetic

date = np.datetime64('2024-03-15')
one_week = np.timedelta64(7, 'D')

print("Date + 1 week:\n", date + one_week)
# Output: Date + 1 week:
# 2024-03-22
print("\n\n")

datetime = np.datetime64('2024-03-15T10:30:00')
two_hours = np.timedelta64(2, 'h')

print("Datetime + 2 hours:\n", datetime + two_hours)
# Output: Datetime + 2 hours:
# 2024-03-15T12:30:00
print("\n\n")

date2 = np.datetime64('2024-03-22')

print("Date2 - Date:\n", date2 - date)  # Difference between dates
# Output: Date2 - Date:
# 7 days
print("\n\n")

# 4. Indexing with Datetime

dates = np.array(['2024-03-10', '2024-03-15', '2024-03-20'], dtype='datetime64')

print("Dates after 2024-03-12:\n", dates[dates > np.datetime64('2024-03-12')])
# Output: Dates after 2024-03-12:
# ['2024-03-15' '2024-03-20']
print("\n\n")

# 7. Working with Time Deltas

time_delta = np.timedelta64(1, 'D')  # 1 day
print("Time delta:\n", time_delta)
# Output: Time delta:
# 1 days
print("\n\n")

time_delta_seconds = np.timedelta64(3600, 's')  # 1 hour in seconds
print("Time delta in seconds:\n", time_delta_seconds)
# Output: Time delta in seconds:
# 3600 seconds
print("\n\n")

# 9. Vectorized Operations with Datetimes

dates = np.array(['2024-03-10', '2024-03-15', '2024-03-20'], dtype='datetime64')
one_day = np.timedelta64(1, 'D')

print("Dates + 1 day:\n", dates + one_day) # Vectorized operation
# Output: Dates + 1 day:
# ['2024-03-11' '2024-03-16' '2024-03-21']
print("\n\n")
```
