### Searching and Sorting 

```python
import numpy as np

arr = np.array([5, 2, 8, 1, 9, 3])

# I- Sorting Functions

print("np.sort(arr): ", np.sort(arr))  # Returns a sorted copy
# Output: np.sort(arr):  [1 2 3 5 8 9]
print("\n\n")

print("np.argsort(arr): ", np.argsort(arr))  # Returns indices that would sort the array
# Output: np.argsort(arr):  [3 1 5 0 2 4]
print("\n\n")

arr2d = np.array([[5, 2, 8], [1, 9, 3]])
print("np.lexsort((arr2d[:, 2], arr2d[:, 0])):\n", np.lexsort((arr2d[:, 2], arr2d[:, 0])))  # Lexicographical sort
# Output: np.lexsort((arr2d[:, 2], arr2d[:, 0])):
# [1 0]
print("\n\n")

arr_in_place = arr.copy()  # Create a copy to sort in-place
arr_in_place.sort()  # Sorts the array in place
print("arr.sort(): ", arr_in_place)
# Output: arr.sort():  [1 2 3 5 8 9]
print("\n\n")

print("np.partition(arr, 3): ", np.partition(arr, 3))  # Partitions around the 3rd smallest element
# Output: np.partition(arr, 3):  [3 2 1 5 8 9] (order of elements before partition point may vary)
print("\n\n")

print("np.argpartition(arr, 3): ", np.argpartition(arr, 3)) # Returns indices that would partition the array
# Output: np.argpartition(arr, 3):  [5 1 3 0 2 4] (order of elements before partition point may vary)
print("\n\n")



# II- Searching Functions

print("np.argmax(arr): ", np.argmax(arr))  # Index of the maximum value
# Output: np.argmax(arr):  4
print("\n\n")

print("np.argmin(arr): ", np.argmin(arr))  # Index of the minimum value
# Output: np.argmin(arr):  3
print("\n\n")

arr_nan = np.array([5, 2, np.nan, 1, 9, 3])
print("np.nanargmax(arr_nan): ", np.nanargmax(arr_nan))  # Index of max, ignoring NaNs
# Output: np.nanargmax(arr_nan):  4
print("\n\n")

print("np.nanargmin(arr_nan): ", np.nanargmin(arr_nan))  # Index of min, ignoring NaNs
# Output: np.nanargmin(arr_nan):  3

print("\n\n")

print("np.where(arr > 4): ", np.where(arr > 4))  # Indices where condition is True
# Output: np.where(arr > 4):  (array([0, 2, 4]),)  (returns a tuple of arrays)
print("\n\n")

arr_sorted = np.array([1, 3, 5, 7, 9])
print("np.searchsorted(arr_sorted, 4): ", np.searchsorted(arr_sorted, 4))  # Index to insert 4
# Output: np.searchsorted(arr_sorted, 4):  2
print("\n\n")

print("np.searchsorted(arr_sorted, [0, 4, 10]): ", np.searchsorted(arr_sorted, [0, 4, 10]))  # Indices to insert multiple values
# Output: np.searchsorted(arr_sorted, [0, 4, 10]):  [0 2 5]
print("\n\n")
```
