# **1. Introduction to Seaborn**

**I. Overview and Installation**
* Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn is particularly well-suited for visualizing complex datasets and for creating visualizations that are aesthetically pleasing and easy to understand.

* **Installation**



```python
!pip install seaborn
```

**II.Comparison with Matplotlib**
* While Matplotlib is a powerful and flexible library for creating a wide range of visualizations, it requires a lot of customization to produce polished plots. Seaborn simplifies this process by providing a high-level interface with sensible defaults, making it easier to generate complex visualizations with minimal code.

**III.Key Differences:**

* Ease of Use: Seaborn provides more intuitive functions for common statistical plots.
* Aesthetics: Seaborn comes with built-in themes and color palettes to enhance the look of plots.
* Data Handling: Seaborn works well with pandas DataFrames, making it easy to plot data directly from these structures.
* Statistical Visualization: Seaborn includes functions for visualizing statistical relationships, distributions, and more, which would require additional coding in Matplotlib.

**Example:**
* A simple comparison of a basic plot in Matplotlib and Seaborn:

* **Simple plot in Matplotlib**

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({'x': [1, 2, 3, 4], 'y': [10, 20, 25, 30]})

plt.plot(data['x'], data['y'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Matplotlib Plot')
plt.show()
```

* **Simaple Plot in Seaborn:**

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({'x': [1, 2, 3, 4], 'y': [10, 20, 25, 30]})

sns.lineplot(data=data, x='x', y='y')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Seaborn Plot')
plt.show()
```

# **2. Seaborn Basics**

**I. Basic Plots**

* **sns.lineplot(x, y, data=None, hue=None, style=None, markers=True, dashes=False, ...)** :Creates a line plot with options to group and style lines by different variables.

* Key Parameters:
    * x, y: Variables to plot.
    * data: DataFrame containing the data.
    * hue: Grouping variable that will produce lines with different colors.
    * style: Grouping variable that will produce lines with different styles.
    * markers: Whether to show markers.
    * dashes: Whether to show dashes.

* Use Case:
   * Visualizing trends over time, such as stock prices.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'value': [10, 15, 13, 17, 19]
})

sns.lineplot(x='day', y='value', data=data)
plt.title('Daily Values')
plt.show()
```

* **sns.scatterplot(x, y, data=None, hue=None, style=None, size=None, ...)**
: Creates a scatter plot with options to vary point colors, shapes, and sizes.

* Key Parameters:
    * x, y: Variables to plot.
    * data: DataFrame containing the data.
    * hue: Grouping variable that will produce points with different colors.
    * style: Grouping variable that will produce points with different shapes.
    * size: Grouping variable that will produce points with different sizes.
* Use Case:
    * Examining the relationship between two continuous variables, such as height and weight.

```python
data = pd.DataFrame({
    'height': [150, 160, 170, 180, 190],
    'weight': [50, 60, 65, 80, 90]
})

sns.scatterplot(x='height', y='weight', data=data)
plt.title('Height vs Weight')
plt.show()
```

* **sns.barplot(x, y, data=None, hue=None, estimator=np.mean, ci='sd', ...)**
: Creates a bar plot with options to display error bars.

* Key Parameters:
    * x, y: Variables to plot.
    * data: DataFrame containing the data.
    * hue: Grouping variable that will produce bars with different colors.
    * estimator: Function to calculate the statistic (e.g., mean, median).
    * ci: Size of the confidence interval to draw around the estimate.

* Use Case:
    * Comparing average values across different categories, such as average test scores by gender.

```python
data = pd.DataFrame({
    'category': ['A', 'B', 'C'],
    'value': [4, 7, 5]
})

sns.barplot(x='category', y='value', data=data)
plt.title('Category Values')
plt.show()
```

**II. Plot Aesthetics and Customization**

* **sns.set()**:Sets global plotting parameters to control the style and scale of plots.

* **sns.set_style(style=None, rc=None)**:Sets the overall aesthetic style of the plots.

* Key Parameters:
    * style: Options include 'darkgrid', 'whitegrid', 'dark', 'white', and 'ticks'.
    * rc: A dictionary of rc parameters to customize the style.

* Use Case:

* Key Parameters:

**IV. Setting Themes**

* **sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)** : Sets the theme for plots with various customization options.

* Key Parameters:
    * context: Controls the scale of plot elements (e.g., 'notebook', 'paper', 'talk', 'poster').
    * style: Sets the overall look and feel (e.g., 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks').
    * palette: Defines the color palette (e.g., 'deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind').
    * font: Sets the font family.
    * font_scale: Adjusts font size relative to the default.
    * color_codes: Maps shorthand color codes (e.g., 'b', 'g', 'r') to palette colors.
    * rc: A dictionary of rc parameters to override other settings.

* Use Case:
    * Applying a consistent theme across all plots in a project.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'category': ['A', 'B', 'C', 'D'],
    'value': [10, 25, 15, 30]
}
data = pd.DataFrame(data)  # Create a pandas DataFrame

sns.set_theme(context='talk', style='whitegrid', palette='muted', font='serif', font_scale=1.2, color_codes=True)

# Now you can use the DataFrame 'data' in your plots
sns.barplot(x='category', y='value', data=data)
plt.title('Bar Plot with Custom Theme')
plt.show()
```

* **Pandas DataFrames:**

    * **Primary Structure:** Seaborn primarily uses Pandas DataFrames, which are two-dimensional, size-mutable, and potentially heterogeneous tabular data structures with labeled axes (rows and columns).

* **Data Formats:**

    * **Long-Form Data:**
        * **Description:** Also known as "tidy" data, where each variable is a column, and each observation is a row.
        * **Usage:** Most Seaborn functions prefer this format.
        * **Example:** For visualizing the relationship between multiple variables.

    * **Wide-Form Data:**
        * **Description:** Each column represents a different variable.
        * **Usage:** Some Seaborn plots, like heatmaps, can use this format.

# **10.Multi-Plot Grids**

**FacetGrid**
* FacetGrid is a powerful tool for visualizing conditional relationships across a grid of subplots. It is particularly useful for visualizing different subsets of data.

* Key Parameters:**
    * data: The DataFrame containing the data to plot.
    * row: Categorical variable that will determine the rows.
    * col: Categorical variable that will determine the columns.
    * hue: Categorical variable that will determine the color.
    * col_wrap: Limits the number of column facets.
    * sharex, sharey: Share the x and y axes across plots.
    * height: Height of each subplot.
    * aspect: Aspect ratio of each subplot.
    * palette: Colors to use for different levels of the hue variable.

* Use Case:
    * creates a grid of histograms of total_bill split by time and sex, with different colors for smokers and non-smokers.


```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="sex", hue="smoker")
g.map(plt.hist, "total_bill")
plt.show()
```

**PairGrid**
* PairGrid is used for plotting pairwise relationships in a dataset. It allows you to visualize the relationship between two variables for different subsets of the data.

* Key Parameters:
    * data: The DataFrame containing the data to plot.
    * hue: Variable in data to map plot aspects to different colors.
    * palette: Colors to use for different levels of the hue variable.
    * vars: Variables to include in the pair plot.
    * x_vars, y_vars: Variables to plot on the x and y axes.
    * diag_sharey: Whether to share the y axis across diagonal subplots.
    * height: Height of each subplot.
    * aspect: Aspect ratio of each subplot.
    * corner: If True, dont add the upper triangle subplots.

* Use Case:
    *  creates a grid of histograms on the diagonal and scatter plots on the off-diagonal, colored by sex.


```python
g = sns.PairGrid(tips, hue="sex")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

**JointGrid**
* JointGrid is used for visualizing a single relationship along with univariate marginal distributions on the same plot. It is ideal for exploring the relationship between two continuous variables.

* Key Parameters:
    * x, y: Variables that specify positions on the x and y axes.
    * data: DataFrame containing the data to plot.
    * kind: Type of plot to draw; can be 'scatter', 'kde', 'hist', etc.
    * height: Size of the figure (it will be square).
    * ratio: Ratio of the size of the marginal axes to the main axes.
    * space: Space between the joint and marginal axes.

* Use Case:
    * displays a scatter plot with a regression line between total_bill and tip, with histograms on the margins.

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.regplot, sns.histplot)
plt.show()
```

# **12. Statistical Estimation**
**I. Error Bars**
  * Description: Error bars represent the variability of data and are commonly used to show the uncertainty of a measurement.
  * Usage: Many Seaborn plots like barplot, pointplot, and lineplot include error bars by default.
  * Customization: Control error bars with parameters like ci (confidence interval), capsize, and errcolor.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.barplot(x='day', y='total_bill', data=tips, ci='sd', capsize=0.2, errcolor='gray')
plt.show()
```

**II. Confidence Intervals**
  * Description: Confidence intervals provide a range of values that are likely to contain the true value of an unknown population parameter.
  * Usage: Seaborn automatically computes confidence intervals for many plot types.
  * Customization: The ci parameter controls the confidence level (default is 95%).

```python
sns.pointplot(x='day', y='total_bill', data=tips, ci=68)  # 68% confidence interval
```

**III. Bootstrapping**
  * Description: Bootstrapping is a statistical method that involves resampling with replacement to estimate the sampling distribution of a statistic.
  * Usage in Seaborn: Functions like barplot and pointplot use bootstrapping to estimate confidence intervals.
  * Customization: The n_boot parameter specifies the number of bootstrap samples.

```python
sns.pointplot(x='day', y='total_bill', data=tips, ci=95, n_boot=2000)
```
